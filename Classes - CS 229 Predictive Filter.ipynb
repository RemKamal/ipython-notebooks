# Predictive Filter  

---  
Since early sensory pathways are typically modeled with linear-nonlinear models [10, 11], I propose to analytically derive the optimal predictive filter of the retina given the statistics of natural scenes - i.e., derive the filter that maximizes the predictive information between the input and output of the LN model.  I will then compare these derived filters with actual data from tiger salamander retinas, using reverse correlation to fit filters to the data.  I will also compare how the optimal filter changes as a function of the scene statistics, to demonstrate whether the functional properties of real retinal circuits are designed or adapted to specific regimes of visual scene (spatial and temporal) correlations.  Other directions could include looking at how the optimal filter varies with input noise (i.e. contrast), and whether in high SNR regimes we recover traditional results of maximizing information to obtain center-surround receptive fields.  


Finding these optimal filters and comparing them with filters learned from data not only provides a test of the predictive information hypothesis, but also could potentially yield important results for how to pre-process sensory information in artificial neural networks.

Simulation of "naturalistic" Gaussian noise  

- Increasing $\tau$ in exponential form results in more correlated noise and an FFT with larger low frequency peaks that drops off faster with frequency  
- The best 1/f fit is obtained by the $-2\ln(|t_2 - t_1)$ covariance formulation

Taken from http://www.pdmi.ras.ru/EIMI/2011/STMP/presentations/Fyodorov.pdf for aperiodic Gaussian stimuli with 1/f spectrum  

Aperiodic version: random Gaussian Fourier integral deﬁnes a Gaussian process on the whole line $-\infty < t < \infty$ by
$$
V(t) = \int_0^\infty \frac{d \omega}{\sqrt{\omega}} [e^{i \omega t} v(\omega) + e^{i \omega t} \overline{v}(\omega)]
$$
and  
$$
\langle V(t_1) V(t_2) \rangle = -2 \ln (|t_2 - t_1|).
$$

```python
n   = 1000
tau = 5
mu  = np.zeros((n,))
cov = np.zeros((n,n))
for i in xrange(n):
    for j in xrange(n):
        if i == j:
            cov[i,j] = 1
        else:
            #cov[i,j] = -2*np.log(np.abs(2*np.sin(np.float(i-j)/2)))
            #cov[i,j] = np.exp(-np.abs(j-i)/tau)
            cov[i,j] = -2*np.log(np.abs(i-j))   # taken from http://www.pdmi.ras.ru/EIMI/2011/STMP/presentations/Fyodorov.pdf for aperiodic Gaussian stimuli with 1/f spectrum
            
        
x = np.random.multivariate_normal(mu,cov)
```

```python
fig = gcf()
fig.set_size_inches((16,5))
plot(x)
```

```python
y = np.fft.fft(x)
```

```python
fig = gcf()
fig.set_size_inches((16,5))
plot(np.abs(y[0:len(y)/2]))
f = linspace(1,500,500)
plot(f,3500/f,'r--')
ylim((0,1200))
title('FFT of Gaussian with covariance -2*log(|i-j|)',fontsize=18)
```

Trying Surya's guess of
$$
\langle V(t_1) V(t_2) \rangle = \frac{1}{\exp(\frac{-|i-j|}{\tau})}.
$$  

- This doesn't fit as well as the $-2 \ln(|t_2 - t_1|)$ that we tried above  
- For various values of $\tau$ tried (2, 5, 10, 15, 20, 50), no good fit was obtained  
- Probably the best-case was $\tau = 15$.  
- The larger the $\tau$, the larger the initial amplitudes are, and the faster the FFT drops off as a function of freq.

```python
n   = 1000
tau = 20
mu  = np.zeros((n,))
cov = np.zeros((n,n))
for i in xrange(n):
    for j in xrange(n):
        if i == j:
            cov[i,j] = 1
        else:
            #cov[i,j] = -2*np.log(np.abs(2*np.sin(np.float(i-j)/2)))
            cov[i,j] = np.exp(-np.abs(j-i)/tau)
            #cov[i,j] = -2*np.log(np.abs(i-j))
        
x = np.random.multivariate_normal(mu,cov)
```

```python
fig = gcf()
fig.set_size_inches((16,5))
plot(x)
```

```python
y = np.fft.fft(x)
```

```python
fig = gcf()
fig.set_size_inches((16,5))
plot(np.abs(y[0:len(y)/2]))
f = linspace(1,500,500)
plot(f,1000/f,'r--')
ylim((0,200))
title('FFT of Gaussian with covariance np.exp(-np.abs(j-i)/tau, tau=10',fontsize=18)
```

Trying periodic covariance model also mentioned by Fyodorov:  
$$
-2 \ln \left ( \left | \, 2\sin \left (\frac{i-j}{2} \right ) \right | \right )
$$

```python
n   = 1000
tau = 5
mu  = np.zeros((n,))
cov = np.zeros((n,n))
for i in xrange(n):
    for j in xrange(n):
        if i == j:
            cov[i,j] = 1
        else:
            cov[i,j] = -2*np.log(np.abs(2*np.sin(np.float(i-j)/2)))
            #cov[i,j] = np.exp(-np.abs(j-i)/tau)
            #cov[i,j] = -2*np.log(np.abs(i-j))   # taken from http://www.pdmi.ras.ru/EIMI/2011/STMP/presentations/Fyodorov.pdf for aperiodic Gaussian stimuli with 1/f spectrum
            
        
x = np.random.multivariate_normal(mu,cov)
```

```python
fig = gcf()
fig.set_size_inches((16,5))
plot(x)
```

```python
y = np.fft.fft(x)
```

```python
fig = gcf()
fig.set_size_inches((16,5))
plot(np.abs(y[0:len(y)/2]))
f = linspace(1,500,500)
plot(f,3500/f,'r--')
ylim((0,1200))
title('FFT of Gaussian with covariance -2*np.log(np.abs(2*np.sin(np.float(i-j)/2)))',fontsize=18)
```
