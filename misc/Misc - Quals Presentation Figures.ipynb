```python
import numpy as np
```

```python
fig = gcf()
fig.set_size_inches((10,6))

ax = subplot()
kandel_pages = [1020, 1080, 1181, 1414, 1760]
plot(kandel_pages,'k',linewidth=5)
for tick in ax.yaxis.get_major_ticks():
    tick.label.set_fontsize(16)
ylabel('Pages',fontsize=25)
```

```python
plot(np.exp(-np.linspace(0,10,2000)),'k',linewidth=4)
```

```python
fig = gcf()
fig.set_size_inches((10,6))

alpha = 3.
theta = 7.
plot(np.linspace(-1,1,2000),1./(1+np.exp(-alpha*(np.linspace(0,10,2000) - theta))),'k',linewidth=4)
vlines(0,-0.2,1.2,colors='k',linestyle='--',linewidth=1.5)
ylim(-0.02,1.02)
```
