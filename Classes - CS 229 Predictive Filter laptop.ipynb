## Load functions for mutual information, generating linear kernel, and estimating (non)regularized error

```python
def linearKernel(frequency,phase,variance,resolution):
    # linearKernel generates the linear filter of an LN model typical of the retina
    # This filter is the dot product of a sinusoid and a gaussian pdf.
    # INPUTS: frequency (of the sinusoid), phase (of the sinusoid), variance (of the gaussian), resolution (how many samples do you want?)
    # OUTPUTS: row vector of the kernel

    x = linspace(0,2*pi,resolution);

    sinusoid = sin(frequency*x + phase);
    samples  = sqrt(variance)*randn(10e3,1);
    gauss    = np.histogram(samples,resolution);

    kernel = sinusoid*gauss[0];
    kernel = kernel/max(kernel);
    
    return kernel
```

```python
def mutiN(voltage, current, nBins, minV, maxV, minC, maxC):
    '''Function to compute the mutual information between neuron voltage and the input current - with just two bins!
    
    Note that you need to use asarray(V) and asarray(current) to get time slices instead of per-neuron slices'''
    
    # could potentially ask for min, max, and then use linspace to specify bin edges
    binsA = linspace(minV, maxV, nBins+1)
    binsB = linspace(minC, maxC, nBins+1)
    
    # note that in this function's MATLAB counterpart, I explicitly made one bin spiking.
    # here I avoid doing that because I don't force a spike to have a particular voltage,
    # I just call it when it passes threshold on the next iteration
    
    # Counting
    CountsAB, xedges, yedges = histogram2d(squeeze(voltage),squeeze(current), bins=[binsA, binsB])
    CountsA = sum(CountsAB,0) # this sums all the rows, leaving the marginal distribution of voltage
    CountsB = sum(CountsAB,1) # this sums all the cols, leaving the marginal distribution of current
    
    pA  = CountsA.astype(float)/float(sum(CountsA))
    pB  = CountsB.astype(float)/float(sum(CountsB))
    pAB = CountsAB.astype(float)/float(sum(CountsAB))
    
    
    if abs(1 - sum(pA)) > 0.01:
        print 'Probabilities pA do not sum to one ' + str(sum(pA))
    if abs(1 - sum(pB)) > 0.01:
        print 'Probabilities pB do not sum to one ' + str(sum(pB))
    if abs(1 - sum(pAB)) > 0.01:
        print 'Probabilities pAB do not sum to one ' + str(sum(pAB))
    
    
    # Entropies
    HA  = 0.0
    HB  = 0.0
    HAB = 0.0
    
    for prob in pB:
        if prob != 0:
            HB = HB - prob*log2(prob)
    
    for j in xrange(len(pA)):
        if pA[j] != 0:
            HA = HA - pA[j] * log2(pA[j])
            for i in xrange(len(pB)):
                if pAB[i,j] != 0:
                    HAB = HAB - pAB[i,j] * log2(pAB[i,j])
                    
    H = [HA, HB, HAB]
    I = HA + HB - HAB
    
    return H, I
```

```python
def IBerror(coeff,beta):
    
    # Xp and Xf
    T      = 1000
    x_t    = np.random.randn(T)
    x_f    = np.fft.rfft(x_t)[:len(x_t)/2]
    f      = linspace(1,T/2,len(x_f))
    fi     = 1./f
    pink_f = x_f*fi
    pink   = (np.fft.ifft(pink_f)).real
    Xp     = pink[:T/4]
    Xf     = pink[T/4:]
    
    # freq = 1.5, phase = -1, variance = 10, resolution = 40
    kernel = linearKernel(abs(coeff[0])+0.1,coeff[1],10,40)
    Y      = np.convolve(Xp,kernel,'same')
    
    # normalize variance because variance increases entropy
    Xp = Xp/var(Xp)
    Xf = Xf/var(Xf)
    Y  = Y/var(Y)
    
    nBins    = 500
    H, Imem  = mutiN(Xp,Y, nBins, min(Xp), max(Xp), min(Y), max(Y))
    H, Ipred = mutiN(Y,Xf, nBins, min(Y), max(Y), min(Xf), max(Xf))
    
    return Imem - beta*Ipred
```

```python
def IBerror_regularized(coeff,beta,X):
    
    # Xp and Xf
    Xp = X[:len(X)/2]
    Xf = X[len(X)/2:]
    
    kernel = np.squeeze(coeff);
    #print str(coeff.shape) + str(Xp.shape)
    Y      = np.convolve(np.squeeze(Xp),np.squeeze(kernel),'same')
    
    # normalize variance because variance increases entropy
    Xp = Xp/var(Xp)
    Xf = Xf/var(Xf)
    Y  = Y/var(Y)
    
    nBins    = 500
    H, Imem  = mutiN(Xp,Y, nBins, min(Xp), max(Xp), min(Y), max(Y))
    H, Ipred = mutiN(Y,Xf, nBins, min(Y), max(Y), min(Xf), max(Xf))
    
    # return information inefficiency + smoothness penalty
    return Imem - beta*Ipred + (1.0/len(kernel))*sum(diff(kernel)**2)
```

```python
def IBerror_package(coeff,beta,X):
    
    # Xp and Xf
    Xp = X[:len(X)/2]
    Xf = X[len(X)/2:]
    
    kernel = np.squeeze(coeff);
    #print str(coeff.shape) + str(Xp.shape)
    Y      = np.convolve(np.squeeze(Xp),np.squeeze(kernel),'same')
    
    # normalize variance because variance increases entropy
    Xp = Xp/var(Xp)
    Xf = Xf/var(Xf)
    Y  = Y/var(Y)
    
    Imem  = ee.mi(Xp,Y)
    Ipred = ee.mi(Y,Xf)
    
    # return information inefficiency + smoothness penalty
    return Imem - beta*Ipred
```
