```python
import numpy as np
```

```python
x = np.zeros((1000,))
x[200:500] = np.ones((500-200,))
plot(x)
```

```python
fig = gcf()
fig.set_size_inches(16,4)
z = np.fft.fft(x)
plot(z)
xlim((-1,200))
```

```python
import matplotlib.mlab as mlab

mean = 0
variance = 1
sigma = sqrt(variance)
y = np.linspace(-3,3,100)
plot(y,mlab.normpdf(y,mean,sigma))
```

```python
y2 = mlab.normpdf(y,mean,sigma)
y2.shape
y3 = -y2
```

```python
w = np.zeros((1000,))
w[200:300] = y2
w[650:750] = y3
```

```python
plot(w)
ylim((-1,1))
```

```python
fig = gcf()
fig.set_size_inches(16,5)
z2 = np.fft.fft(w)
plot(z2)
xlim((0,200))
```
