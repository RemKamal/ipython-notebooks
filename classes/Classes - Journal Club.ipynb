#### Read csv files and parse ratings

```python
import re
numbers = re.compile(r'(\d+)')
def numericalSort(value):
    parts = numbers.split(value)
    parts[1::2] = map(int, parts[1::2])
    return parts

prefix = 'csv/'
ratings = []
for infile in sorted(glob.glob(os.path.join(prefix, '*')), key=numericalSort):
    print "Current File Being Processed is: " + infile
    headers = np.genfromtxt(open(infile,'r'),delimiter=',',dtype='string')[0]
    ratings.append(np.array(np.genfromtxt(open(infile,'r'),delimiter=',',dtype='f8')[1:]))
```
