```python
import numpy as np
from scipy.linalg import toeplitz
from scipy.linalg import inv
from numpy import fft
import brewer2mpl
```

```python
def pinkNoise(duration):
    x_t    = np.random.randn(2*duration)
    x_f    = np.fft.rfft(x_t)[:len(x_t)/2]
    f      = linspace(1,duration,len(x_f))
    fi     = 1./f
    pink_f = x_f*fi

    return real(np.fft.ifft(pink_f))
```

```python
x = pinkNoise(1000)
plot(x)
```

```python
# generate upper triangular toeplitz matrix from x
tmp = zeros(len(x))
tmp[0] = x[0]
toe = toeplitz(tmp,x)
```

```python
bmap = brewer2mpl.get_map('Greys', 'sequential', 7)
imshow(np.outer(x,x), cmap=bmap.mpl_colormap)
colorbar()
```

```python
def getW(x,y):
    
    #if len(y.shape) < 2:
    #    y = np.reshape(y,(len(y),1))
    
    return np.dot(inv(np.dot(np.transpose(x),x)), np.dot(np.transpose(x),y))
```

Need # rows of X to be equal to number of rows in y.  This means that the number of past x's has to be equal to the number of past y's

X = [[x_t x_t-1 x_t-2 ... x_t-n], [x_t+1 x_t x_t-1 x_t-2 ... x_t-(n-1)], [x_t+2, ..., x_t-(n-2)], ..., [x_t+m, ..., x_t]]

```python
n = 200     #length of filter = number of times you need in the past
m = 100000   #number of predictions = number of times you need in the future

x_all  = pinkNoise(2*m+n)
x_past = x_all[:(m+n-1)]
y      = x_all[(m+n-1):-1]

first_row = x_past[:n]
first_row = first_row[::-1] #reverse the first row, since it's x_t, x_t-1, ..., x_t-n+1
first_col = x_past[n-1:]  #this should go from n to m but it's weird since python doesn't count the first element

toe = toeplitz(first_col,first_row)

W = getW(toe,y)
```

```python
fig = gcf()
fig.set_size_inches(17,5)

subplot(1,3,1)
bmap = brewer2mpl.get_map('Greys', 'sequential', 7)
imshow(np.dot(np.transpose(toe),toe), cmap=bmap.mpl_colormap)
colorbar()
title('Inner Product X^T X')

subplot(1,3,2)
imshow(inv(np.dot(np.transpose(toe),toe)), cmap=bmap.mpl_colormap)
colorbar()
title('Inverse of Inner Product X^T X')

subplot(1,3,3)
#imshow(np.dot(np.transpose(toe), np.reshape(y,(len(y),1))), cmap=bmap.mpl_colormap)
#colorbar()
plot(range(1,n+1),np.dot(np.transpose(toe), np.reshape(y,(len(y),1))))
xlim(1,50)
title('Inner product X^T Y')
```

```python
plot(getW(toe,y),'k')
xlim(-5,n+5)
title('W', fontsize=14)
ax = gca()
ax.set_xlim(ax.get_xlim()[::-1])
xlabel('Time before present', fontsize=14)

name = "example filter, n%d, m%d.eps" % (n,m)
savefig(name)
```

### We don't want X and Y to be disjoint.  Rather Y should be a copy of X shifted by $n$

Here x[n] = y[0].  This means that x_t+1 = y_t+1.  To see this, note that toe[1,0] (x_t+1) = y[0].

```python
n = 200     #length of filter = number of times you need in the past
m = 100000   #number of predictions = number of times you need in the future

x_all  = pinkNoise(m+n+1)
x      = x_all[:(m+n-1)]
y      = x_all[n:-1]

first_row = x[:n]
first_row = first_row[::-1] #reverse the first row, since it's x_t, x_t-1, ..., x_t-n+1
first_col = x[n-1:]  #this should go from n to m but it's weird since python doesn't count the first element

toe = toeplitz(first_col,first_row)

W = getW(toe,y)
```

```python
from numpy import sin, linspace, pi
from pylab import plot, show, title, xlabel, ylabel, subplot
from scipy import fft, arange

def plotSpectrum(y,Fs):
 """
 Plots a Single-Sided Amplitude Spectrum of y(t)
 """
 n = len(y) # length of the signal
 k = arange(n)
 T = n/Fs
 frq = k/T # two sides frequency range
 frq = frq[range(n/2)] # one side frequency range

 Y = fft(y)/n # fft computing and normalization
 Y = Y[range(n/2)]
 
 plot(frq,abs(Y),'r') # plotting the spectrum
 xlabel('Freq (Hz)')
 ylabel('|Y(freq)|')

Fs = 1000.0;  # sampling rate
Ts = 1.0/Fs; # sampling interval
t = arange(0,len(W)*Ts,len(W)) # time vector

subplot(2,1,1)
plot(W, 'k')
xlim(-5,n+5)
title('W', fontsize=14)
ax = gca()
ax.set_xlim(ax.get_xlim()[::-1])
xlabel('Time before present', fontsize=14)

subplot(2,1,2)
plotSpectrum(W,Fs)
```

#### The delta function shifts the power of the entire FFT spectrum.  Increasing just the delta function magnitude, we increase the power of the whole FFT (right) while keeping the same shape.

```python
fig = gcf()
fig.set_size_inches(15,5)

subplot(1,2,1)
tmp = W
tmp[0] = tmp[0]*10
plot(tmp)
xlim(-5,200+5)

subplot(1,2,2)
tmp_freq = fft(tmp)
plot(tmp_freq, 'k')
title('FFT of W', fontsize=14)
xlabel('Frequency', fontsize=14)
ylabel('Power', fontsize=14)
```

### Increasing contrast at constant mean

```python
n     = 200      #length of filter = number of times you need in the past
m     = 100000   #number of predictions = number of times you need in the future
SNR   = 0.1
slope = 0.0001
intercept = 0

t      = np.linspace(SNR,slope*(m+n+1),m+n+1)
x_all  = intercept + t*randn(m+n+1)
x      = x_all[:(m+n-1)]
y      = x_all[n:-1]

first_row = x[:n]
first_row = first_row[::-1] #reverse the first row, since it's x_t, x_t-1, ..., x_t-n+1
first_col = x[n-1:]  #this should go from n to m but it's weird since python doesn't count the first element

toe = toeplitz(first_col,first_row)

W = getW(toe,y)
```
