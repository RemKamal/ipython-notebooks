```python
import numpy as np
from scipy.linalg import toeplitz
from scipy.linalg import inv
from numpy import fft
import brewer2mpl
```

```python
def pinkNoise(duration):
    x_t    = np.random.randn(2*duration)
    x_f    = np.fft.rfft(x_t)[:len(x_t)/2]
    f      = linspace(1,duration,len(x_f))
    fi     = 1./f
    pink_f = x_f*fi

    return real(np.fft.ifft(pink_f))
```

```python
x = pinkNoise(1000)
plot(x)
```

```python
# generate upper triangular toeplitz matrix from x
tmp = zeros(len(x))
tmp[0] = x[0]
toe = toeplitz(tmp,x)
```

```python
bmap = brewer2mpl.get_map('Greys', 'sequential', 7)
imshow(np.outer(x,x), cmap=bmap.mpl_colormap)
colorbar()
```

```python
def getW(x,y):
    
    #if len(y.shape) < 2:
    #    y = np.reshape(y,(len(y),1))
    
    return np.dot(inv(np.dot(np.transpose(x),x)), np.dot(np.transpose(x),y))
```

Need # rows of X to be equal to number of rows in y.  This means that the number of past x's has to be equal to the number of past y's

X = [[x_t x_t-1 x_t-2 ... x_t-n], [x_t+1 x_t x_t-1 x_t-2 ... x_t-(n-1)], [x_t+2, ..., x_t-(n-2)], ..., [x_t+m, ..., x_t]]

```python
n = 200     #length of filter = number of times you need in the past
m = 100000   #number of predictions = number of times you need in the future

x_all  = pinkNoise(2*m+n)
x_past = x_all[:(m+n-1)]
y      = x_all[(m+n-1):-1]

first_row = x_past[:n]
first_row = first_row[::-1] #reverse the first row, since it's x_t, x_t-1, ..., x_t-n+1
first_col = x_past[n-1:]  #this should go from n to m but it's weird since python doesn't count the first element

toe = toeplitz(first_col,first_row)

W = getW(toe,y)
```

```python
fig = gcf()
fig.set_size_inches(17,5)

subplot(1,3,1)
bmap = brewer2mpl.get_map('Greys', 'sequential', 7)
imshow(np.dot(np.transpose(toe),toe), cmap=bmap.mpl_colormap)
colorbar()
title('Inner Product X^T X')

subplot(1,3,2)
imshow(inv(np.dot(np.transpose(toe),toe)), cmap=bmap.mpl_colormap)
colorbar()
title('Inverse of Inner Product X^T X')

subplot(1,3,3)
#imshow(np.dot(np.transpose(toe), np.reshape(y,(len(y),1))), cmap=bmap.mpl_colormap)
#colorbar()
plot(range(1,n+1),np.dot(np.transpose(toe), np.reshape(y,(len(y),1))))
xlim(1,50)
title('Inner product X^T Y')
```

```python
plot(getW(toe,y),'k')
xlim(-5,n+5)
title('W', fontsize=14)
ax = gca()
ax.set_xlim(ax.get_xlim()[::-1])
xlabel('Time before present', fontsize=14)

name = "example filter, n%d, m%d.eps" % (n,m)
savefig(name)
```
