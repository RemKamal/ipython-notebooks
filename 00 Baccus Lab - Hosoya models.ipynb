### Anti-Hebbian retina model

```python
def lightResponse(stimulus, nCells=16, nRGCs=1, tau=100.0, beta=5.0):
    ''' Function to return retinal ganglion cell light response.
    
    Input:
        Stimulus (ndarray) of type (nCells x time)
        
        nCells (int) that specifies the number of amacrine and bipolar cells
        
        tau (float) that specifies the plasticity of amacrine synapses
    '''
    cellsOnEdge = int(np.ceil(np.sqrt(nCells)))
    sample_RF = np.zeros((cellsOnEdge, cellsOnEdge))
    sample_RF[1:3,1:3] = 0.5
    B = array([sample_RF.reshape((sample_RF.shape[0]*sample_RF.shape[1],))]*nRGCs)    # bipolar-RGC synapses
    A = 0.5*B # amacrine-RGC synapses
    #C = np.cov(stimulus.transpose())
    
    duration = stimulus.shape[1]
    response = []
    As = []
    
    for t in xrange(duration):
        C = np.outer(stimulus[:,t],stimulus[:,t].transpose())
        A = A + -(1.0/tau)*(A + beta*np.dot((A + B),C))
        As.append(A)
        response.append(np.dot(B+A,stimulus[:,t]))
    
    return response, As
```

### So it seems that we have the approximate parameters.  Now let's see if we can reproduce their stimulus sequence and get the same results

```python
def checkerboard(w, h, d=0):
    re = np.r_[ w/2*[0,1] ]
    ro = np.r_[ w/2*[1,0] ]
    if d==0:
        return np.row_stack(h/2*(re, ro))
    else:
        board = np.row_stack(h/2*(re, ro))
        return np.dstack(d/2*(board, board))
```

```python
def largeCheckers(w, h, d=0):
    re = np.r_[ w/4*[0], w/2*[1], w/4*[0] ]
    ro = np.r_[ w/4*[1], w/2*[0], w/4*[1] ]
    if d==0:
        return np.row_stack(h/4*(re, ro, ro, re))
    else:
        board = np.row_stack(h/4*(re, ro, ro, re))
        return np.dstack(d/2*(board, board))
```

```python
def bars(w, h, d=0, orientation='vertical'):
    cross_section = np.r_[ w/4*[0], w/4*[1,1], w/4*[0] ]
    if orientation=='vertical':
        bar = vstack(h/2*(cross_section, cross_section))
    elif orientation=='horizontal':
        bar = hstack(h/2*(np.hsplit(cross_section, w), np.hsplit(cross_section, w)))
    if d==0:
        return bar
    else:
        return np.dstack(d/2*(bar, bar))
```

#### Note that using a checkerboard with checkers of size one pixel doesn't work, since then the excitation will always be zero, and the As just have to be uniform.

```python
def generateStimuli(cellsPerSide=4, timePerCondition=1000):
    nConditions = 7
    stim = np.zeros((cellsPerSide, cellsPerSide, nConditions*timePerCondition))
    randn1d = randn(1,1,timePerCondition)
    randn41 = np.row_stack(2*(randn1d,randn1d))
    board   = (2*largeCheckers(cellsPerSide,cellsPerSide) - 1).reshape((cellsPerSide,cellsPerSide,1))
    vbars   = (2*bars(cellsPerSide,cellsPerSide,orientation='vertical') - 1).reshape((cellsPerSide,cellsPerSide,1))
    hbars   = (2*bars(cellsPerSide,cellsPerSide,orientation='horizontal') - 1).reshape((cellsPerSide,cellsPerSide,1))
    
    stim[:,:,  timePerCondition:2*timePerCondition] = randn(cellsPerSide,cellsPerSide,timePerCondition) # Random White Checkers
    stim[:,:,2*timePerCondition:3*timePerCondition] = np.hstack(2*(randn41,randn41)) # Uniform flicker
    stim[:,:,3*timePerCondition:4*timePerCondition] = np.kron(randn(1,1,timePerCondition), board) # Checkerboard flickering
    stim[:,:,4*timePerCondition:5*timePerCondition] = np.kron(randn(1,1,timePerCondition), vbars) # Vertical bars flickering
    stim[:,:,5*timePerCondition:6*timePerCondition] = np.kron(randn(1,1,timePerCondition), hbars) # Horizontal bars flickering
    
    return stim
```

### First let's put a threshold in the light response

```python
def lightResponse_N(stimulus, nCells=16, nRGCs=1, tau=100.0, beta=5.0):
    ''' Function to return retinal ganglion cell light response.
    
    Input:
        Stimulus (ndarray) of type (nCells x time)
        
        nCells (int) that specifies the number of amacrine and bipolar cells
        
        tau (float) that specifies the plasticity of amacrine synapses
    '''
    cellsOnEdge = int(np.ceil(np.sqrt(nCells)))
    sample_RF   = np.zeros((cellsOnEdge, cellsOnEdge))
    sample_RF[1:3,1:3] = 0.5
    B = array([sample_RF.reshape((sample_RF.shape[0]*sample_RF.shape[1],))]*nRGCs)    # bipolar-RGC synapses
    A = 0.5*B # amacrine-RGC synapses
    
    threshold = 0.6    
    duration = stimulus.shape[1]
    response = []
    As = []
    
    for t in xrange(duration):
        C = np.outer(stimulus[:,t],stimulus[:,t].transpose())
        A = A + -(1.0/tau)*(A + beta*np.dot((A + B),C))
        As.append(A)
        response.append(np.dot(B+A,stimulus[:,t]))
    
    response = response/np.max(response)
    
    for t in xrange(len(response)):
        if response[t] < threshold:
            response[t] = 0
        else:
            response[t] = response[t]*(response[t] - threshold)
    
    return response, As
```

## Next let's implement the optimal "predictive" MMSE estimate of the future given the past by setting $A = C_{xy} C_y^{-1}$ where $x$ is the future and $y$ is the past.  

The optimal predictor is given by $\hat{x} = C_{xy} C_y^{-1} (y - E[y]) + E[x]$.

```python
def lightResponse_pred(stimulus, nCells=16, nRGCs=1, tau=100.0, beta=5.0):
    ''' Function to return retinal ganglion cell light response.
    
    Input:
        Stimulus (ndarray) of type (nCells x time)
        
        nCells (int) that specifies the number of amacrine and bipolar cells
        
        tau (float) that specifies the plasticity of amacrine synapses
    '''
    cellsOnEdge = int(np.ceil(np.sqrt(nCells)))
    sample_RF   = np.zeros((cellsOnEdge, cellsOnEdge))
    sample_RF[1:3,1:3] = 0.5
    B = array([sample_RF.reshape((sample_RF.shape[0]*sample_RF.shape[1],))]*nRGCs)    # bipolar-RGC synapses
    A = 0.5*B # amacrine-RGC synapses
    
    #threshold = 0.6    
    duration = stimulus.shape[1]
    response = []
    As = []
    
    for t in xrange(1,duration):
        Cy  = np.outer(stimulus[:,t],stimulus[:,t].transpose()) # Cpast
        Cyi = linalg.inv(Cy + (1.0/5.0)*np.max(Cy)*np.eye(Cy.shape[0])) # with ridge regression since may be singular otherwise
        Cxy = np.outer(stimulus[:,t],stimulus[:,t-1].transpose()) # Cpast-future
        
        A = A + -(1.0/tau)*(A + beta*np.dot((A + B),np.dot(Cxy,Cyi)))
        #A = A - (1.0/tau)*(A + np.dot(Cxy,Cyi))
        As.append(A)
        response.append(np.dot(B+A,stimulus[:,t]))
    
    response = response/np.max(response)
    
#     for t in xrange(len(response)):
#         if response[t] < threshold:
#             response[t] = 0
#         else:
#             response[t] = response[t]*(response[t] - threshold)
    
    return response, As
```

### A seems to be doing the right thing in general, but the weights blow up quickly (2500/7000 frames).  Increasing tau fixes the blow-up problem, but then the filters no longer look like good predictions at all.

Part of the problem is that we're being unclear about what we're predicting.  Maybe we should actually be predicting the future excitation, $Bx_t$.

```python
def lightResponse_pred(stimulus, nCells=16, nRGCs=1, tau=100.0, beta=5.0):
    ''' Function to return retinal ganglion cell light response.
    
    Input:
        Stimulus (ndarray) of type (nCells x time)
        
        nCells (int) that specifies the number of amacrine and bipolar cells
        
        tau (float) that specifies the plasticity of amacrine synapses
    '''
    cellsOnEdge = int(np.ceil(np.sqrt(nCells)))
    sample_RF   = np.zeros((cellsOnEdge, cellsOnEdge))
    sample_RF[1:3,1:3] = 0.5
    B = array([sample_RF.reshape((sample_RF.shape[0]*sample_RF.shape[1],))]*nRGCs)    # bipolar-RGC synapses
    A = 0.5*B # amacrine-RGC synapses
    
    #threshold = 0.6    
    duration = stimulus.shape[1]
    response = []
    As = []
    
    for t in xrange(1,duration):
        #Cy  = np.outer(stimulus[:,t],stimulus[:,t].transpose()) # Cpast
        #Cyi = linalg.inv(Cy + (1.0/5.0)*np.max(Cy)*np.eye(Cy.shape[0])) # with ridge regression since may be singular otherwise
        #Cxy = np.outer(stimulus[:,t],stimulus[:,t-1].transpose()) # Cpast-future
        #A = A + -(1.0/tau)*(A + beta*np.dot((A + B),np.dot(Cxy,Cyi)))
        #A = A - (1.0/tau)*(A + np.dot(Cxy,Cyi))
        #A = A - (1.0/tau)*(A + beta*np.dot(Cxy,Cyi))
        
        # Want A to subtract a prediction of the excitatory pathways
        Cy  = np.outer(stimulus[:,t-1], stimulus[:,t-1]) # covariance of past excitation Bx
        Cyi = linalg.inv(Cy + (1.0/5.0)*np.max(Cy)*np.eye(Cy.shape[0]))
        Cxy = np.outer(np.dot(B,stimulus[:,t]), stimulus[:,t-1]) # covariance between future excitation and past excitation
        
        W          = np.dot(Cxy,Cyi)
        A_response = np.dot(W, stimulus[:,t] - np.mean(stimulus[:,:t], axis=1)) + np.mean(np.dot(B, stimulus[:,:t]))
        A          = np.dot(A_response, linalg.pinv(stimulus[:,t].reshape((stimulus[:,t].shape[0],1))))
           
        As.append(A)
        response.append(np.dot(B,stimulus[:,t]) - A_response)
    
    response = response/np.max(response)
    
#     for t in xrange(len(response)):
#         if response[t] < threshold:
#             response[t] = 0
#         else:
#             response[t] = response[t]*(response[t] - threshold)
    
    return response, As
```

```python
def lightResponse_pred(stimulus, nCells=16, nRGCs=1, tau=100.0, beta=5.0):
    ''' Function to return retinal ganglion cell light response.
    
    Input:
        Stimulus (ndarray) of type (nCells x time)
        
        nCells (int) that specifies the number of amacrine and bipolar cells
        
        tau (float) that specifies the plasticity of amacrine synapses
    '''
    cellsOnEdge = int(np.ceil(np.sqrt(nCells)))
    sample_RF   = np.zeros((cellsOnEdge, cellsOnEdge))
    sample_RF[1:3,1:3] = 0.5
    B = array([sample_RF.reshape((sample_RF.shape[0]*sample_RF.shape[1],))]*nRGCs)    # bipolar-RGC synapses
    A = 0.5*B # amacrine-RGC synapses
    
    #threshold = 0.6    
    duration = stimulus.shape[1]
    response = []
    As = []
    Ws = []
    
    for t in xrange(1,duration):
        #Cy  = np.outer(stimulus[:,t],stimulus[:,t].transpose()) # Cpast
        #Cyi = linalg.inv(Cy + (1.0/5.0)*np.max(Cy)*np.eye(Cy.shape[0])) # with ridge regression since may be singular otherwise
        #Cxy = np.outer(stimulus[:,t],stimulus[:,t-1].transpose()) # Cpast-future
        #A = A + -(1.0/tau)*(A + beta*np.dot((A + B),np.dot(Cxy,Cyi)))
        #A = A - (1.0/tau)*(A + np.dot(Cxy,Cyi))
        #A = A - (1.0/tau)*(A + beta*np.dot(Cxy,Cyi))
        
        # Want A to subtract a prediction of the excitatory pathways
        Cy  = np.outer(stimulus[:,t-1], stimulus[:,t-1]) # covariance of past stimulus
        Cyi = linalg.inv(Cy + (1.0/5.0)*np.max([np.max(Cy), 0.1])*np.eye(Cy.shape[0]))
        Cxy = np.outer(np.dot(B,stimulus[:,t]), stimulus[:,t-1]) # covariance between future excitation and past excitation
        
        W          = np.dot(Cxy,Cyi)
        A_response = np.dot(W, stimulus[:,t] - np.mean(stimulus[:,:t], axis=1)) + np.mean(np.dot(B, stimulus[:,:t]))
        A          = np.dot(A_response, linalg.pinv(stimulus[:,t].reshape((stimulus[:,t].shape[0],1))))
           
        As.append(A)
        Ws.append(W)
        response.append(np.dot(B,stimulus[:,t]) - A_response)
    
    #response = response/np.max(response)
    
#     for t in xrange(len(response)):
#         if response[t] < threshold:
#             response[t] = 0
#         else:
#             response[t] = response[t]*(response[t] - threshold)
    
    return response, As, Ws
```

### Sunrise/sunset experiment

```python
def sunrise(cellsPerSide=4, slope=1.0):
    frames  = int(255.0/slope)
    randn1d = (255 - 127)*0.15*randn(1,1,frames)
    randn41 = np.row_stack(2*(randn1d,randn1d))
    stim    = np.hstack(int(cellsPerSide/2.0)*(randn41,randn41)) # Uniform flicker
    
    for t in xrange(frames):
        mean = t*1.0
        stim[:,:,t] = stim[:,:,t] + mean
        
    stim = stim.astype(int)
    stim[stim > 255] = 255
    stim[stim < 0]   = 0
    
    return stim
```

```python
stim_brighter = sunrise()
r_sunrise_antiHebb, As_sunrise_antiHebb = lightResponse(stim_brighter.reshape((4*4,stim_brighter.shape[2]))/float(np.max(stim_brighter)), nCells=16, tau=100.0, beta=5.0)
r_sunrise_predict, As_sunrise_predict, Ws_sunrise   = lightResponse_pred(stim_brighter.reshape((4*4,stim_brighter.shape[2])), nCells=16, tau=10.0, beta=5.0)
```
