```python
import numpy as np
import pandas as pd
from scipy import fft
```

```python
def plotSpectrum(y,Fs=1000.0,colorstr='r',fillin=True,transparency=0.7,plt=True):
 '''
 Plots a Single-Sided Amplitude Spectrum of y(t).
 
 Example:
 Fs = 1000.0;  # sampling rate
 Ts = 1.0/Fs; # sampling interval
 t = arange(0,len(W)*Ts,len(W)) # time vector

 subplot(2,1,1)
 plot(W, 'k')
 xlim(-5,n+5)
 title('W', fontsize=14)
 ax = gca()
 ax.set_xlim(ax.get_xlim()[::-1])
 xlabel('Time before present', fontsize=14)

 subplot(2,1,2)
 plotSpectrum(W,Fs)
 '''
 n = len(y) # length of the signal
 k = arange(n)
 T = n/Fs
 frq = k/T # two sides frequency range
 frq = frq[range(n/2)] # one side frequency range

 Y = fft(y)/n # fft computing and normalization
 Y = Y[range(n/2)]
 
 if plt:
     plot(frq,abs(Y),color=colorstr,alpha=transparency) # plotting the spectrum
     if fillin:
            fill_between(frq, abs(Y), 1e-6, color=colorstr, alpha=0.4)
     xlabel('Freq (Hz)', fontsize=14)
     ylabel('|Y(freq)|', fontsize=14)
     xlim(-10,frq[-1])
 else:
     return frq, abs(Y)
```

```python
def negProfit(lam,stock_prices,fit_line,budget=100000.):

    fit_line = np.exp(np.log(fit_line)+lam)
    
    # Assuming you can buy fractional shares
    shares = [(fit_line[idx]-price)/price for idx,price in enumerate(stock_prices) if price < fit_line[idx]]
    purchases = [(fit_line[idx]-price) for idx,price in enumerate(stock_prices) if price < fit_line[idx]]
    
    if sum(shares) < 100.:
        return 0.0
    else:
        return -(((data['Close'][0]*sum(shares) - sum(purchases))/(sum(purchases)+budget))**(1./64.) - 1)
```

### Define model.

```python
def runModel(lam, alpha_d, beta_d, alpha_p, beta_p, monthly_shareprice, monthly_household_income):
    x_mo  = np.linspace(0,len(monthly_shareprice),len(monthly_shareprice))
    ps_mo = np.polyfit(x_mo,np.log(monthly_shareprice),deg=1)
    fit_line_mo = np.exp(ps_mo[0]*x_mo + ps_mo[1] + lam)
    
    bank = 0.
    shares = 0.
    for m, price in enumerate(monthly_shareprice):
        # receive your paycheck
        bank = bank + 0.1*monthly_household_income[m]

        # check how expensive/cheap the market is
        # positive if below watermark
        discount = (fit_line_mo[m]- price)/price

        # allocate how much of bank you want to invest
        if discount >= 0.0:
            purchase = (alpha_d*discount + beta_d)*bank
        else:
            purchase = (alpha_p*discount + beta_p)*bank
        if purchase > bank:
            purchase = bank
        elif purchase < 0.0:
            purchase = 0.0

        # buy shares
        shares = shares + purchase/price
        bank   = bank - purchase

        # accrue interest in bank
        bank   = bank + (0.01/12.)*bank

    final_price  = list(monthly_shareprice)[-1]
    # total_return
    return (bank + shares*final_price - 0.1*sum(monthly_household_income))/(0.1*sum(monthly_household_income))
```

```python
def negModel(w, monthly_shareprice, monthly_household_income):
    lam = w[0]
    alpha_d = w[1] # multiplicative bias towards buying shares in cheap market
    beta_d  = w[2] # additive bias towards buying shares in cheap market
    alpha_p = w[3]
    beta_p  = w[4]
    
    total_return = runModel(lam, alpha_d, beta_d, alpha_p, beta_p, monthly_shareprice, monthly_household_income)
    
    return -total_return
```

```python
lam = res.x[0]
alpha_d = res.x[1] # multiplicative bias towards buying shares in cheap market
beta_d  = res.x[2] # additive bias towards buying shares in cheap market
alpha_p = res.x[3]
beta_p  = res.x[4]

x_mo  = np.linspace(0,len(mo_shareprice_sub),len(mo_shareprice_sub))
fit_line_mo = np.exp(np.log(mo_GDP_sub) + lam)

bank = 0.
shares = 0.
purchase_history = []
bank_history = []
fraction_history = []
discount_history = []
for m, price in enumerate(mo_shareprice_sub):
    # receive your paycheck
    bank = bank + 0.1*mo_household_sub[m]
    
    # check how expensive/cheap the market is
    # positive if below watermark
    discount = (fit_line_mo[m]- price)/price
    discount_history.append(discount)
    
    # allocate how much of bank you want to invest
    if discount >= 0.0:
        purchase = (alpha_d*discount + beta_d)*bank
    else:
        purchase = (alpha_p*discount + beta_p)*bank
    if purchase > bank:
        purchase = bank
    elif purchase < 0.0:
        purchase = 0.0
    
    # buy shares
    shares = shares + purchase/price
    bank   = bank - purchase
    
    # accrue interest in bank
    #bank   = bank + 0.01*bank
    purchase_history.append(purchase)
    bank_history.append(bank)
    fraction_history.append(purchase/monthly_household_income[m])

final_price  = list(mo_shareprice_sub)[-1]
total_return = (bank + shares*final_price - 0.1*sum(mo_household_sub))/(0.1*sum(mo_household_sub))
```
