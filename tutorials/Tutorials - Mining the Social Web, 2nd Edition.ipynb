## Example 1. Using boilerpipe to extract text from a web page

```python
from boilerpipe.extract import Extractor
```

```python
URL='http://www.fool.com/investing/general/2014/06/20/apple-how-the-beats-acquisition-fits-into-the-big.aspx?source=ihpsitth0000001'
extractor = Extractor(extractor='ArticleExtractor', url=URL)
print extractor.getText()
```

## Example 2. Using feedparser to extract the text (and other fields) from an RSS or Atom feed

```python
import feedparser
```

```python
FEED_URL='http://feeds2.feedburner.com/MotleyFoolMoney'
fp = feedparser.parse(FEED_URL)

for e in fp.entries:
    print e.title
    print e.links[0].href
    print e.summary_detail.value
```

## Example 3. NLTK tool kit

```python
import nltk
import re
nltk.download('punkt')
```

```python
txt = fp.entries[0].summary_detail.value
txt = re.sub(r'(\<.*?\>)|(\n)|(&nbsp;)', '', txt)
print txt
```

```python
sentences = nltk.tokenize.sent_tokenize(txt)
print sentences
```

```python
tokens = [nltk.tokenize.word_tokenize(s) for s in sentences]
print tokens
```
