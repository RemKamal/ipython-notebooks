```python
import numpy as np
import mdp
```

```python
X = np.random.randn(1000,2)
Z = np.random.randn(1000,1)
X[:,1] = np.squeeze(0.8*X[:,0:1] + 0.1*Z)
```

```python
Y = mdp.pca(X)
```

```python
fig = gcf()
fig.set_size_inches(16,6)
subplot(121)
scatter(Y[:,0],Y[:,1])
subplot(122)
scatter(X[:,0],X[:,1])
```
