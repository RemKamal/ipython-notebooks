### (4) Spike triggered covariance

```python
# Generate new spike train with quadratic nonlinearity
# stim
dt = 2.   # ms
T = 10e5 # ms
time = np.linspace(0,T,T/dt+1)
s    = randn(len(time))

# filter
filter_len  = 50. # ms
tau         = 10  # ms
omega       = 0.3 # rad/ms
filter_time = np.linspace(0,filter_len,filter_len/dt+1)
filt = np.array([np.exp(-t/tau) * np.sin(omega*t) for t in filter_time]) # linear filter

# linear response
u = np.convolve(s,filt,'same')

# nonlinear response
# nonlinearity
theta = 15.
delta = 2.
r_max = 0.1 # was 20.
r_u   = np.array([r_max/(1.+np.exp((theta - u[t]**2)/delta)) for t in xrange(len(u))])
plot(time,r_u,'k')
```

```python
# want average firing rate to be around 20 Hz
ideal_firing_rate_per_bin = 20*dt*(1./1000.)
#ideal_firing_rate_per_bin = 20.
(ideal_firing_rate_per_bin, np.mean(r_u))
```

```python
# generate spike train
spiketrain = poisson(r_u)
```

```python
plot(time,spiketrain,'k.')
g = ylim((-0.1,max(spiketrain)+0.1))
```

```python
# STA
sta = [u[t-len(filt):t] for t in xrange(len(filt),len(s)) if spiketrain[t] > 0]
sta = np.mean(sta,axis=0)
sta = sta[::-1]

fig, ax1 = plt.subplots()
ax1.plot(filter_time, sta, 'k')

ax2 = ax1.twinx()
ax2.plot(filter_time, filt, 'k--')
annotate('True filter',(3,-0.2),fontsize=12)
annotate('STA',(17,0.1),fontsize=12)
title('STA versus real filter')
```

#### STA can't recover the filter of a quadratic nonlinearity because the nonlinearity is non-monotonic.  This means that the STA is averaging positive and negative stim values together, cancelling them out.

```python
data = []
prior = []
for idt,sp in enumerate(spiketrain):
    if idt > len(filt):
        prior.append(s[idt-len(filt):idt])
        if sp > 0:
            data.append(s[idt-len(filt):idt])
            
data  = np.array(data)
data  = data.transpose()
prior = np.array(prior)
prior = prior.transpose()
shape(data)
```

#### The STC successfully retrives the model because we are only averaging squared values of the stimulus, and so we don't average out the two opposing signed stimulus that both elicit spikes.

```python
# two-dimensional model
# stim
dt = 2.   # ms
T = 10e5 # ms
time = np.linspace(0,T,T/dt+1)
s    = randn(len(time))

# filter
filter_len  = 50. # ms
tau_f       = 10  # ms
tau_g       = 30  # ms
omega_f     = 0.3 # rad/ms
omega_g     = 0.2 # rad/ms
filter_time = np.linspace(0,filter_len,filter_len/dt+1)
f = np.array([np.exp(-t/tau_f) * np.sin(omega_f*t) for t in filter_time]) # linear filter
g = np.array([-np.exp(-t/tau_g) * np.cos(omega_g*t) for t in filter_time])

# linear response
u_f = np.convolve(s,f,'same')
u_g = np.convolve(s,g,'same')

# nonlinear response
# nonlinearity
theta = 15.
delta = 2.
r_max = 95000.
#r_max = 20.
r_u   = np.array([r_max/((1.+np.exp((theta - u_f[t])/delta))*(1+np.exp((theta - u_g[t])/delta))) for t in xrange(len(u_f))])
plot(time,r_u,'k')
```

```python
# want average firing rate to be around 20 Hz
ideal_firing_rate_per_bin = 20*dt*(1./1000.)
#ideal_firing_rate_per_bin = 20.
(ideal_firing_rate_per_bin, np.mean(r_u))
```

```python
fig = gcf()
fig.set_size_inches((10,4))

# sketch r in the space (u_f,u_g)
uf_range = np.linspace(np.min(u_f),np.max(u_f),100)
ug_range = np.linspace(np.min(u_g),np.max(u_g),100)
uf_range = uf_range[::-1]
rs = zeros((100,100))
for idf,fi in enumerate(uf_range):
    for idg,gi in enumerate(ug_range):
        rs[idf,idg] = r_max/((1.+np.exp((theta - fi)/delta))*(1+np.exp((theta - gi)/delta)))

imshow(rs,interpolation='nearest',extent=[np.min(ug_range), np.max(ug_range), np.min(uf_range), np.max(uf_range)])
xlabel('u_g',fontsize=12)
ylabel('u_f',fontsize=12)
colorbar()
```

## (5) GLM

#### (a) Generate white noise

```python
dt = 2.   # ms
T = 10e5 # ms
time = np.linspace(0,T,T/dt+1)
s    = randn(len(time))
```

#### (b) Generate spike history filters

```python
postspike_filter_time = np.linspace(0,30,30./dt)
tmp = np.append(0.2*ones(10),np.linspace(0.2,0,len(postspike_filter_time)-10))
refractory = -np.sin(0.2*postspike_filter_time) * np.exp(-0.2*postspike_filter_time)
#refractory = refractory
bursting   = -np.sin(0.2*postspike_filter_time + 0.2) * np.exp(-0.1*postspike_filter_time) + tmp
#bursting   = 2.*bursting - 1
bursting   = bursting + (1.0 - sum(bursting))/len(bursting)

plot(postspike_filter_time,refractory,'g')
plot(postspike_filter_time,bursting,'b')
```

```python
# Linear component
f = np.array([np.exp(-t/tau_f) * np.sin(omega_f*t) for t in filter_time]) # linear filter
lambda_0 = .025 # baseline firing rate
u_f = np.convolve(s,f,'same') + np.log(lambda_0)

plot(time,np.exp(u_f),'k')
```

```python
prefilter_spiking = poisson(np.exp(u_f))
plot(time,prefilter_spiking,'k.')
```

```python
firingrate = sum(prefilter_spiking[0:500]) # numSpikes in a sec = Hz
firingrate
```

```python
postfilter_spiking_ref = poisson(np.exp(u_f)*np.exp(np.convolve(prefilter_spiking,refractory,'same')))
postfilter_spiking_bur = poisson(np.exp(u_f)*np.exp(np.convolve(prefilter_spiking,bursting,'same')))
plot(time,postfilter_spiking_ref)
firingrate = sum(postfilter_spiking_ref[0:500]) # numSpikes in a sec = Hz
firingrate
```

```python
def raster(spikes,color='k',dt=1.):
    ax = gca()
    for ith, sp in enumerate(spikes):
        if sp > 0:
            vlines(ith*dt, .5, 1.5, color=color,linewidth=1.3)
            #ylim(.5, len(spikes) + .5)
    
    return ax
```

```python
fig = gcf()
fig.set_size_inches((20,10))

subplot(3,1,1)
ax = raster(prefilter_spiking[:500],dt=2.)
title('Before post-spike filter',fontsize=14)
xlabel('Time (ms)')

subplot(3,1,2)
ax = raster(postfilter_spiking_ref[:500],dt=2.)
title('After refractory post-spike filter',fontsize=14)
xlabel('Time (ms)')

subplot(3,1,3)
ax = raster(postfilter_spiking_bur[:500],dt=2.)
title('After bursting post-spike filter',fontsize=14)
xlabel('Time (ms)')

tight_layout()
```
