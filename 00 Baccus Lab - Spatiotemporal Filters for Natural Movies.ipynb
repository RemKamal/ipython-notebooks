## Import standard predictive filter functions

```python
def getW(x,y,mode='auto'):
    ''' Solve XW = Y where X is toeplitz, y is a vector.
    
    1) 'auto' mode uses numpy's builtin lstsq solver to avoid
    taking matrix inversions or multiplying by illconditioned matrices.
    2) 'direct' mode uses direct matrix inversion by xtx and returns a 
    warning when xtx is ill-conditioned.
    '''
    
    if mode=='direct':
        xtx = np.dot(np.transpose(x),x)
        con = np.linalg.cond(xtx)
        if con > 10e3:
            print('Condition number of XTX is very high, %f, and inverse wont be accurate.') % (con)
        W = np.dot(np.linalg.inv(xtx), np.dot(np.transpose(x),y))
    elif mode=='auto':
        W, res, rank, s = np.linalg.lstsq(x, y, rcond=-1)
    else:
        print('mode can be either auto or direct')
    
    return W
```

```python
def toe_block(x,height,width,pixels_per_n):
    '''x is a vector, height is dimension m = # of predictions,
    and width is dimension n = length of filter.
    '''
    m = height
    n = width
    toe = np.zeros((m,n*pixels_per_n))
    for idx in xrange(m):
        toe[idx,:] = x[idx*pixels_per_n:(idx+1)*pixels_per_n + (n-1)*pixels_per_n]
        toe[idx,:] = toe[idx,:][::-1]
    
    return toe
```

### Now we have a list of flattened frames.  How do we transform this into our desired Toeplitz matrix $X$?

```python
shape(movie[0])
```

```python
movie_flat = np.array(movie).flat
```

```python
n = 20
pixels_per_n = shape(movie[0])[0]
lag = 1
m = nFrames - n - lag
```

```python
(np.mean(movie_flat), np.var(movie_flat))
```

```python
x = movie_flat[:(m+n)*pixels_per_n]
y = np.array(center_pixel) #movie_flat[n*pixels_per_n:]
y = y[n+1:]
```

```python
X_past = toe_block(x,m,n,pixels_per_n)
```
