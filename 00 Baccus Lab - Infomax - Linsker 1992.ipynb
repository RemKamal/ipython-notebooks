```python
def varianceConstraint(C,Q,eta):
    allNodes = diag((C.dot(Q)).dot(C.transpose()) + eta*C.dot(C.transpose()))
    return sum([abs(1 - v) for v in allNodes])
```

```python
def varianceFreqConstraint(q,c):
    if len(q)!=len(c):
        print 'q and c are not the same length!'
    else:
        K = len(q)
        return (1./K)*sum([(q[k] + eta)*(c[k]**2) for k in range(K)])
```

```python
def callbackFn(mu):
    #print varianceFnMu(mu)
    print mu
```

```python
def uniqueSolution(eta,beta,q,mode='manual'):
    K = len(q)
    
    def getFilter(mu):
        return np.array([np.sqrt((beta/eta)*max([0,(-1 + (qk/(2*(qk+eta)))*(1 + np.sqrt(1-(2*eta*K)/(mu*beta*qk))))])) for qk in q])
    
    if mode=='auto':
        def varianceFnMu(mu):
            c = getFilter(mu)
            return 10*abs(1. - varianceFreqConstraint(q,c))

        mu  = -abs(randn(1))
        res = minimize(varianceFnMu, x0=mu, method='Nelder-Mead', callback=callbackFn)

        muOpt = res.x
        return getFilter(muOpt)
    else:
        return getFilter
```
