```python
entropy = []
for i in xrange(len(fig[0])):
    total = sum(fig[0][i])
    tmp = 0
    for j in xrange(len(fig[0])):
        p = np.max([np.float(fig[0][i][j])/total, 1e-10])
        tmp = tmp - p*log2(p)
    entropy.append(tmp)
```
