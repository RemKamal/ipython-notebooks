```python
fig, ax1 = plt.subplots()
fig.set_size_inches((10,8))
t = [0,1,2]
bears = ['father','mother','baby']
ax1.plot(t, [60,40,50], 'k-')
ax1.plot(t, [19.5, 18.4, 20.01], 'k--')
ax1.set_xlabel('Porridges',fontsize=16)
plt.xticks(t, bears, fontsize=14)
# Make the y-axis label and tick labels match the line color.
ax1.set_ylabel('Temperature (C)', color='k',fontsize=16)
for tl in ax1.get_yticklabels():
    tl.set_color('k')


ax2 = ax1.twinx()
ax2.plot(t, [15, 10, 3], 'b-')
ax2.set_ylabel('Volume (oz)', color='b', fontsize=16)
for tl in ax2.get_yticklabels():
    tl.set_color('b')


fill_between([0,1,2],[15,10,3],0,color='0.8',alpha=0.3)
annotate('Measured Ambient Temperature', [0.2, 1.9], fontsize=14)
grid()

savefig('complicated.png')
```
