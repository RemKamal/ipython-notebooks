```python
def Srec(st,mt):
    return np.abs(st*(mt**2))
```

```python
def S(st,mt,mn):
    return np.abs((Srec(st,mt))*mn)
```

```python
def N(ne,mn):
    ni = 10e-4
    return np.abs(ne*mn + ni)
```

```python
def constraint(mn,st,ne):
    return 3.0 - np.sqrt(sum(S(st,mt,mn)+N(ne,mn)))
```

```python
def infoTransferRate(lam,st,mt,ne):
    neural_filter = mn(st,mt,ne,ni,lam)
    return sum(np.log(1.0 + S(st,mt,neural_filter)/N(ne,neural_filter)))
```

```python
coeff = np.abs(1 - np.linspace(-1,1,150)**2) #np.ones(Nt/2.0) #result_in
res   = optimize.minimize(infoTransferRate, x0=coeff, args=(st,ne), method='SLSQP', 
                          constraints={'type' : 'ineq', 'fun' : constraint, 'args' : [st,ne]}, 
                          tol=1e-6, options={'disp': True})
```

```python
plot(res.x)
```

```python
def mn(st,mt,ne,ni,lam):
    num = -ni*(2*ne + Srec(st,mt)) + np.sqrt(np.abs((ni**2)*(Srec(st,mt)**2) - 4*ne*Srec(st,mt)*ni*(1.0/lam)))
    den = 2*ne*(ne+Srec(st,mt))
    
    return num/den
```

#### How do we reconcile van Hateren's finding that the filter maximizes frequencies near SNR = 1 and attenuates strong SNR frequencies with the theory of matched filters and my results in maximizing information?

```python
def neg_infoTransferRate(lam,st,mt,ne):
    neural_filter = mn(st,mt,ne,ni,lam)
    return -sum(np.log(1.0 + S(st,mt,neural_filter)/N(ne,neural_filter)))
```

```python
lam = 10.0 #result_in
res1  = optimize.minimize(neg_infoTransferRate, x0=lam, args=(st,mt,ne), method='nelder-mead', 
                          tol=1e-6, options={'disp': True})
```
