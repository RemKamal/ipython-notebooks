```python
def toe(x,height,width):
    '''x is a vector, height is dimension m = # of predictions,
    and width is dimension n = length of filter.
    '''
    m = height
    n = width
    first_row = x[:n]
    first_row = first_row[::-1] #reverse the first row, since it's x_t, x_t-1, ..., x_t-n+1
    first_col = x[n-1:]  #this should go from n to m but it's weird since python doesn't count the first element

    return toeplitz(first_col,first_row)
```

```python
def getW(x,y,mode='auto'):
    ''' Solve XW = Y where X is toeplitz, y is a vector.
    
    1) 'auto' mode uses numpy's builtin lstsq solver to avoid
    taking matrix inversions or multiplying by illconditioned matrices.
    2) 'direct' mode uses direct matrix inversion by xtx and returns a 
    warning when xtx is ill-conditioned.
    '''
    
    if mode=='direct':
        xtx = np.dot(np.transpose(x),x)
        con = np.linalg.cond(xtx)
        if con > 10e3:
            print('Condition number of XTX is very high, %f, and inverse wont be accurate.') % (con)
        W = np.dot(np.linalg.inv(xtx), np.dot(np.transpose(x),y))
    elif mode=='auto':
        W, res, rank, s = np.linalg.lstsq(x, y, rcond=-1)
    else:
        print('mode can be either auto or direct')
    
    return W
```

```python
def pinkNoise(duration):
    x_t    = np.random.randn(2*duration)
    x_f    = np.fft.rfft(x_t)[:len(x_t)/2]
    f      = linspace(1,duration,len(x_f))
    fi     = 1./f
    pink_f = x_f*fi

    return real(np.fft.ifft(pink_f))
```

```python
def gaussianInfoSNR(w,x,x_in):
    signal = np.convolve(x,w,'valid')
    noise  = np.convolve(x_in-x,w,'valid')
    
    n = len(signal) # length of the signal

    S = fft(signal)/n # fft computing and normalization #-np.mean(signal)
    S = abs(S[range(n/2)])
    S = np.convolve(S,(1.0/500.0)*np.ones(500),'same')
    N = fft(noise)/n
    N = abs(N[range(n/2)])
    N = np.convolve(N,(1.0/500.0)*np.ones(500),'same')
    
    # SNR(f) is P_s/P_n or (A_s/A_n)**2
    #SNR = [(S[f]/N[f])**2 for f in xrange(n/2)]
    
    return 0.5*np.log2(1+(np.mean(S)/np.mean(N))**2)
```

```python
def negInfo_SNR_smoothed(w,xs,x_ins):
    mi = []
    for t in xrange(len(xs)):
        mi.append(-gaussianInfoSNR(w,xs[t],x_ins[t]))
        
    return np.mean(mi)
```

```python
def negInfo_SNR(w,x,x_in):
    return -gaussianInfoSNR(w,x,x_in)
```

```python
def powerConstraint(w,x_in):
    return 1.0 - np.var(np.convolve(x_in,w))
```

```python
def S(w,x):
    Fs=1000.0
    n1 = len(w)
    n2 = len(x)
    k1 = arange(n1)
    k2 = arange(n2)
    T1 = n1/Fs
    T2 = n2/Fs
    frq1 = k1/T1 # two sides frequency range
    frq2 = k2/T2
    frq1 = frq1[range(n1/2)] # one side frequency range
    frq2 = frq2[range(n2/2)]
    frq1 = np.append(frq1,500.)

    w_frq = fft(w)/len(w)
    w_frq = w_frq[range(len(w)/2)]
    w_frq = np.append(w_frq,w_frq[-1])
    w_frq = interp1d(frq1,w_frq, kind='cubic')
    w_frq = w_frq(frq2)

    x_frq = fft(x)/len(x)
    x_frq = x_frq[range(len(x)/2)]
    
    return np.abs(x_frq*w_frq)
```

```python
def N(w,n):
    Fs=1000.0
    n1 = len(w)
    n2 = len(n)
    k1 = arange(n1)
    k2 = arange(n2)
    T1 = n1/Fs
    T2 = n2/Fs
    frq1 = k1/T1 # two sides frequency range
    frq2 = k2/T2
    frq1 = frq1[range(n1/2)] # one side frequency range
    frq2 = frq2[range(n2/2)]
    frq1 = np.append(frq1,500.)

    w_frq = fft(w)/len(w)
    w_frq = w_frq[range(len(w)/2)]
    w_frq = np.append(w_frq,w_frq[-1])
    w_frq = interp1d(frq1,w_frq, kind='cubic')
    w_frq = w_frq(frq2)
    
    n_frq = fft(n)/len(n)
    n_frq = n_frq[range(len(n)/2)]
    
    return np.abs(n_frq*w_frq)
```

```python
def powerSpectrumConstraint(w,x,n):
    return 1.0 - np.sqrt(sum(S(w,x)+N(w,n)))
```

```python
def negInfoTransferRate(w,x,n):
    return -np.sum(np.log((1+S(w,x))/N(w,n)))
```

```python
def ouProcess(meanCurrent,variance,N):
    ''' shotNoise(meanCurrent, variance, N) '''
    
    dt    = 0.5
    tau   = 3 # ms
    tau2  = N/20
    M     = 30
    t     = linspace(0,M*dt,M)
    t2    = linspace(0,N*dt,N)
    x     = meanCurrent + np.sqrt(variance)*np.random.randn(N+M-1)
    y     = np.exp(-t/tau)
    y2    = np.exp(-t2/tau2)
    z0    = 100
    drift = (z0 - meanCurrent)*y2
    z     = np.convolve(x,y, mode = 'valid')
    z     = z - np.mean(z) + meanCurrent + drift
    #z[0]  = z0
    
    return z
```
