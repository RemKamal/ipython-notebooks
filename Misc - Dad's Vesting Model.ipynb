### Data

```python
current_price = 13.35
current_date  = datetime.datetime.today()
```

```python
# grants vest 25% after year 1, then 1/48 every month thereafter
# rsus vest 1/12 each month
# 10b5 plan automatically sells 3600 shares from $5.2 grants each 
# month given the stock > $15
numGrants   = 4
grant_dates = [datetime.datetime(2012,3,29),
               datetime.datetime(2012,5,11),
               datetime.datetime(2013,2,4),
               datetime.datetime(2014,3,24)]
grant_prices = [5.2, 5.2, 9.83, 18.73]
grant_vested = [4000, 22500, 41250, 4687]
grant_unvested = [62500, 17500, 68750, 51563]
grant_exercised = [83500,0,0,0]
grant_total    = np.array(grant_vested) + np.array(grant_unvested) + np.array(grant_exercised)
rsus = 11718
rsus_vested = 0
```

### The Model

####Assumptions:  
1) The total options (vested + unvested) is the total from which the "fraction-vested-each-month" should be computed.  Note that this contradicts the oldest grant having only 4,000 vested shares (6% of total shares for that grant).  
2) Options vest 25% after year 1, then 1/36 each month afterwards.  Note that this contradicts with the most recent grant already having 4,687 vested options, since that grant is less than a year old.  
3) RSUs vest 1/12 each quarter

```python
vested_options = zeros((numMonths,numGrants)) # cumulative
vested_rsus    = zeros((numMonths,)) # cumulative
total_profit   = zeros((numMonths,))
```

```python
future_dates[8] - grant_dates[3]
```

```python
for idx,d in enumerate(future_dates):
    if idx == 0:
        vested_options[idx,:] = np.array(grant_vested)
        vested_rsus[idx]      = 0
        for g in xrange(numGrants):
            price_difference  = 0.5*((current_price - grant_prices[g]) + np.sqrt((current_price - grant_prices[g])**2))
            total_profit[idx] = total_profit[idx] + price_difference*(vested_options[idx,g])
        total_profit[idx] = total_profit[idx] + current_price*vested_rsus[idx]
    else:
        for g in xrange(numGrants):
            if grant_vested[g] + grant_exercised[g] <= grant_total[g]:
                delta_t = d - grant_dates[g]
                
                # if it's the first month after a year
                if delta_t > datetime.timedelta(365) and delta_t < datetime.timedelta(395) and g < numGrants-1:
                    newly_vested = 0.25*grant_total[g]
                    grant_vested[g] = grant_vested[g] + newly_vested
                    
                # if it's after the first month of the first year but before the grant has fully vested, which should be 1186 days!
                elif (delta_t >= datetime.timedelta(395) and delta_t < datetime.timedelta(2500)) or g >= numGrants-1:
                    newly_vested = (1./48.)*grant_total[g]
                    grant_vested[g] = grant_vested[g] + newly_vested
                    
                    # can't exceed total number of options
                    if grant_vested[g] + grant_exercised[g] >= grant_total[g]:
                        grant_vested[g] = grant_total[g] - grant_exercised[g]
            
            # in this month, the total vested options including the ones that are newly vested
            vested_options[idx,g] = grant_vested[g]
                
            # don't allow negative price_difference
            price_difference  = 0.5*((current_price - grant_prices[g]) + np.sqrt((current_price - grant_prices[g])**2))
            
            # total profit from options
            total_profit[idx] = total_profit[idx] + price_difference*(vested_options[idx,g])
                
        for idq,qtr in enumerate(future_quarters):
            days_until_qtr = qtr - d
            # if less than a month after quarter started, and less than one day before quarter starts
            if days_until_qtr > datetime.timedelta(-30) and days_until_qtr < datetime.timedelta(1):
                if vested_rsus[idx] > 0:
                    msg = 'Multiple months %d think its the beginning of this quarter %d' %(idx,idq)
                    print(msg)
                
                if rsus_vested < rsus:
                    rsus_vested = rsus_vested + (1./12.)*rsus
                    
                # can't exceed total number of rsus
                if rsus_vested >= rsus:
                    rsus_vested = rsus
                
        # in this month, the total vested rsus including ones that are newly vested
        # make sure you count rsus even in months when new ones don't vest
        vested_rsus[idx] = rsus_vested
        # total profit from options and rsus
        total_profit[idx] = total_profit[idx] + vested_rsus[idx]*current_price
```

## Heatmap

```python
def drange(start, stop, step):
    r = [start]
    while r[-1] < stop:
        r.append(r[-1]+step)
    return r
```

```python
prices = drange(10,20,0.5)
```

```python
price_vs_date = zeros((len(prices),len(future_dates)))
```

```python
numMonths = 25 # 29 for until Dec 2016
firstQuarter = datetime.datetime(2014,7,1)
future_dates    = [current_date + relativedelta(months=x) for x in xrange(numMonths)]
future_quarters = [firstQuarter + relativedelta(months=x) for x in xrange(0,numMonths,3)] # a quarter is every 3 months

for idp,p in enumerate(prices):
    vested_options = zeros((numMonths,numGrants)) # cumulative
    vested_rsus    = zeros((numMonths,)) # cumulative
    total_profit   = zeros((numMonths,))
    current_price  = p
    grant_vested = [4000, 22500, 41250, 4687]
    grant_unvested = [62500, 17500, 68750, 51563]
    grant_exercised = [83500,0,0,0]
    grant_total    = np.array(grant_vested) + np.array(grant_unvested) + np.array(grant_exercised)
    rsus = 11718
    rsus_vested = 0
    for idx,d in enumerate(future_dates):
        if idx == 0:
            vested_options[idx,:] = np.array(grant_vested)
            vested_rsus[idx]      = 0
            for g in xrange(numGrants):
                price_difference  = 0.5*((current_price - grant_prices[g]) + np.sqrt((current_price - grant_prices[g])**2))
                total_profit[idx] = total_profit[idx] + price_difference*(vested_options[idx,g])
            total_profit[idx] = total_profit[idx] + current_price*vested_rsus[idx]
        else:
            for g in xrange(numGrants):
                if grant_vested[g] + grant_exercised[g] <= grant_total[g]:
                    delta_t = d - grant_dates[g]

                    # if it's the first month after a year
                    if delta_t > datetime.timedelta(365) and delta_t < datetime.timedelta(395) and g < numGrants-1:
                        newly_vested = 0.25*grant_total[g]
                        grant_vested[g] = grant_vested[g] + newly_vested

                    # if it's after the first month of the first year but before the grant has fully vested, which should be 1186 days!
                    elif (delta_t >= datetime.timedelta(395) and delta_t < datetime.timedelta(2500)) or g >= numGrants-1:
                        newly_vested = (1./48.)*grant_total[g]
                        grant_vested[g] = grant_vested[g] + newly_vested

                        # can't exceed total number of options
                        if grant_vested[g] + grant_exercised[g] >= grant_total[g]:
                            grant_vested[g] = grant_total[g] - grant_exercised[g]

                # in this month, the total vested options including the ones that are newly vested
                vested_options[idx,g] = grant_vested[g]

                # don't allow negative price_difference
                price_difference  = 0.5*((current_price - grant_prices[g]) + np.sqrt((current_price - grant_prices[g])**2))

                # total profit from options
                total_profit[idx] = total_profit[idx] + price_difference*(vested_options[idx,g])

            for idq,qtr in enumerate(future_quarters):
                days_until_qtr = qtr - d
                # if less than a month after quarter started, and less than one day before quarter starts
                if days_until_qtr > datetime.timedelta(-30) and days_until_qtr < datetime.timedelta(1):
                    if vested_rsus[idx] > 0:
                        msg = 'Multiple months %d think its the beginning of this quarter %d' %(idx,idq)
                        print(msg)

                    if rsus_vested < rsus:
                        rsus_vested = rsus_vested + (1./12.)*rsus

                    # can't exceed total number of rsus
                    if rsus_vested >= rsus:
                        rsus_vested = rsus

            # in this month, the total vested rsus including ones that are newly vested
            # make sure you count rsus even in months when new ones don't vest
            vested_rsus[idx] = rsus_vested
            # total profit from options and rsus
            total_profit[idx] = total_profit[idx] + vested_rsus[idx]*current_price
        
    price_vs_date[idp,:] = total_profit
```
