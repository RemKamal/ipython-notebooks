```python
def ensemble(adaptiveIndex, numNeurons, inputType, duration):
    ''' Simulates an ensemble of neurons
    
    inputs: adaptiveIndex, numNeurons, and inputType
    
    possible inputType includes:
        'step'
        'white'
        'ou'
        'shotnoise'
        'brownian'
    
    returns: V (neurons x time), w, spikes, stimulus
    '''
    
    # constants
    N           = duration
    a           = adaptiveIndex
    M           = numNeurons
    v0          = -70.6    # mV
    meanCurrent = 555.0    # pA ?
    variance    = 100.0
    delta       = 0.5
    x0          = 0.0
    
    # initialize variables
    T = np.linspace(0,N*delta,N) # time points corresponding to inputCurrent (same size as V, w, I)
    
    # meanCurrents are chosen for one spike @ a=13 and no spikes at a=18
    
    if inputType == 'step':
        current = stepFn(meanCurrent+70,variance,N,M)
    elif inputType == 'white':
        current = whiteNoise(meanCurrent+50,50*variance,N,M)
    elif inputType == 'ou':
        current = ouProcess(meanCurrent+150,3*variance,N,M)
    elif inputType == 'shotnoise':
        current = shotNoise(meanCurrent+15,variance,N,M)
    elif inputType == 'brownian':
        current = brownian(np.ones((M,)) * meanCurrent, N, delta, variance/25.0, out=None)
        current[:,0] = x0
        
    V,w,spikes,sptimes = aEIF(a, current, v0)
        
    return V, w, spikes, sptimes, T, current
```

```python
# Get ensemble of neurons
V,w,spikes,sptimes,T,stimulus = ensemble(13,100,'step',1000)

# Visualize
fig = gcf()
fig.set_size_inches(12,8)
ax1 = subplot(311)
fig = plot(T,np.array(V).transpose())
ax2 = subplot(312, sharex = ax1)
fig = raster(sptimes)
ax3 = subplot(313, sharex = ax1)
fig = plot(T,np.array(stimulus).transpose())
```

```python
x = asarray(V)
y = asarray(stimulus)
minV = -70 #min(flatten(x))
maxV = 20 #max(flatten(x))
minC = 0#min(flatten(y))
maxC = 700 #max(flatten(y))

info = []
struct = []

nBins = 500  

for i in xrange(shape(x)[1]):
    H, I = mutiN(x[:,i],y[:,i],nBins,minV,maxV,minC,maxC)
    info.append(I)

for i in xrange(shape(y)[1]-1):
    H, I = mutiN(y[:,i],y[:,i+1],nBins,minC,maxC,minC,maxC)
    struct.append(I)
    

fig = gcf()
fig.set_size_inches((16,4))
plot(info)
plot(struct,'r--')
xlim((-10,1000))
```
