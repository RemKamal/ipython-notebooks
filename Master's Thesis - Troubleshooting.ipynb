```python
x = asarray(V)
y = asarray(stimulus)
minV = min(flatten(x))
maxV = max(flatten(x))
minC = min(flatten(y))
maxC = max(flatten(y))

info = []
struct = []

nBins = 500
#minV = min(flatten(x))
#maxV = max(flatten(x))
#minC = min(flatten(y))
#maxC = max(flatten(y))

for i in xrange(shape(x)[1]):
    H, I = mutiN(x[:,i],y[:,i],nBins,minV,maxV,minC,maxC)
    info.append(I)

for i in xrange(shape(y)[1]-1):
    H, I = mutiN(y[:,i],y[:,i+1],nBins,minC,maxC,minC,maxC)
    struct.append(I)
    

fig = gcf()
fig.set_size_inches((16,4))
plot(info)
plot(struct,'r--')
xlim((-10,1000))
```
