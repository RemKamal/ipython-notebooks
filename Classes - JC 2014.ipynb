```python
def nonlinearity(x,K=1.,g=1.,theta=0.,mode='sigmoid',normalized=False):
    if normalized:
        if mode == 'sigmoid':
            n = K/(1+np.exp(-g*(x - theta)))
        elif mode == 'tanh':
            if type(x)==list or type(x)==numpy.ndarray:
                n = [g*tanh(a-theta) + K - g for a in x]
            else:
                n = g*tanh(x-theta) + K - g
        return np.array(n)/np.sqrt(np.var(np.array(n)))
    else:
        if mode == 'sigmoid':
            return K/(1+np.exp(-g*(x - theta)))
        elif mode == 'tanh':
            if type(x)==list or type(x)==numpy.ndarray:
                return [g*tanh(a-theta) + K - g for a in x]
            else:
                return g*tanh(x-theta) + K - g
```

```python
# Return the characteristic matrix
def characteristicMatrix(data,maxGridSize):
    gridSizes = range(2,maxGridSize)
    I = zeros((len(gridSizes),len(gridSizes)))
    M = zeros((len(gridSizes),len(gridSizes)))
    for idx,x in enumerate(gridSizes):
        for idy,y in enumerate(gridSizes):
            I[idx,idy] = maxMI(data,x,y)
            M[idx,idy] = I[idx,idy]/min(np.log(x),np.log(y))
    return M
```

```python
def generateNewPartition(partition,newElement):
    '''partition is a list of bin edges'''
    if newElement in partition:
        return partition
    elif newElement > max(partition):
        partition.append(newElement)
        return partition
    else:
        for idp,p in enumerate(partition):
            if newElement < p:
                partition.insert(idp,newElement)
                return partition
```
