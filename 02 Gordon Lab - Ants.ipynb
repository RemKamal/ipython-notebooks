```python
def importData(verbose):
    myData = 'interactions-486-EC.csv'
    
    # parse data
    headers = np.genfromtxt(open(myData,'r'),delimiter=',',dtype='string')[0]
    dataset = (np.array(np.genfromtxt(open(myData,'r'),delimiter=',',dtype='f8')[1:]))
    #print 'The dataset has size ' + str(dataset.shape)
    
    # initialize constants
    frameRate = 30
    deltaT = 1./frameRate
    
    # Normalize timing information, 
    # convert from frames to seconds, and
    # sort the dataset by the interaction time
    dataset[:,3] = (dataset[:,3] - np.min(dataset[:,3]))*deltaT
    dataset = dataset[dataset[:,3].argsort()]
    
    
    # use ant type (col 6) to only use ants going out (type=2)
    antsLeaving = dataset[dataset[:,5] == 2]

    
    # retrieve unique ant ids
    antsLeavingIDs = np.unique(antsLeaving[:,0])  

    # For each leaving ant, get the timestamps
    # of when it interacts with a returning forager (clickType=2)
    interactionTimes = []
    for ant in antsLeavingIDs:
        # reduce whole dataset to only data pertaining to a single ant
        singleAntLeaving = antsLeaving[antsLeaving[:,0] == ant]
        # use click type (col 5) to only use interactions w Seeded Ants
        singleAntInteractions = singleAntLeaving[singleAntLeaving[:,4] == 2]
        # append the list of interaction timestamps (col 4)
        interactionTimes.append(singleAntInteractions[:,3])
    
    # For each leaving ant, get the timestamps
    # of when it interacts with an ant w/o a seed (clickType=1)
    interactionTimesOther = []
    for ant in antsLeavingIDs:
        # reduce whole dataset to only data pertaining to a single ant
        singleAntLeaving = antsLeaving[antsLeaving[:,0] == ant]
        # use click type (col 5) to only use interactions w Non-Seeded Ants
        singleAntInteractions = singleAntLeaving[singleAntLeaving[:,4] == 1]
        # append the list of interaction timestamps (col 4)
        interactionTimesOther.append(singleAntInteractions[:,3])
    
    # Compute times of departure for ants that decide to forage
    timeOfDeparture = []
    for ant in antsLeavingIDs:
        # reduce whole dataset to only data pertaining to a single ant
        singleAntLeaving = antsLeaving[antsLeaving[:,0] == ant]
        # append the timeOfDeparture timestamp (col 4)
        timeOfDeparture.append(singleAntLeaving[len(singleAntLeaving)-1,3])
    
    if verbose == 1:
        # duration of data in seconds
        print 'Duration of the Data is ' + str(np.max(dataset[:,3])) + ' seconds,'
        print 'there are ' + str(antsLeaving.shape[0]) + ' total interactions with ants that decide to forage,'
        print 'and ' + str(antsLeavingIDs.shape[0]) + ' total ants that decide to forage.'
    
    return [interactionTimes, timeOfDeparture, interactionTimesOther]
```

## Integrate And Forage Model  
$$ \begin{align}
\frac{d x(t)}{dt} &= - \frac{(x(t) - c)}{\tau} + \sigma \, \eta(t) + \sum_i \Delta x \, \delta(t - t_i) \\
\frac{1}{\Delta x} \frac{d x(t)}{dt} &= - \frac{(x(t) - c)}{\tau \Delta x} + \frac{\sigma}{\Delta x} \, \eta(t) + \sum_i \, \delta(t - t_i) \\
\frac{d x'(t)}{dt} &= - \frac{(x'(t) - c)}{\tau'} + \sigma' \, \eta(t) + \sum_i \, \delta(t - t_i)
\end{align}$$

###Current problems:  

- It seems that the returned solution is very close to initialization values.  
   - For instance, using $x0 = [0.01, 0.2, 10, 0.05, 0]$ gives a solution that is less than 0.1 away from any of those values (probably even closer).  This has mean squared error around 15.  
   - And using $x0 = [0.01, 0.2, 40, 0.05, 0]$ gives the solution $ \sim [0.01,   0.2,   39, 0.05, 0]$ with mean squared error of 7.45474975396
   - And using $x0 = [0.01, 40, 0.05, 0]$ (fewer parameters) still gives solution $ \sim [0.01,   39, 0.05, 0]$

To do:  
- Should simplify redundancy in model by dividing by $\Delta x$

```python
def integrateAndForage(interactions,
                       timeOfDeparture,
                       c = 0.01, # internal bias to forage
                       dx = 0.2, # effect size of interaction
                       tau = 10, # time constant of ant's memory
                       sigma = 0.05, # input noise variance
                       sigma2 = 0,
                       threshold = 1): # output noise variance
    # interactions needs to be numpy array
    interactions = np.array(interactions)
    
    # initialize constants and variables
    penalty = 100   # penalty for not deciding to forage
    dt      = 1.0/30.0
    
    if interactions.shape[0] > 0:
        time    = np.linspace(min(interactions),timeOfDeparture,(timeOfDeparture - min(interactions))/dt)
    else:
        time    = np.linspace(0,timeOfDeparture,timeOfDeparture/dt)      
    pred    = np.zeros(len(time))
    noise   = np.random.randn(len(time))
    pred[0] = sigma*noise[0]
    
    # discretize interactions into our sample rate
    # impulse[0] are number of samples, impulse[1] are bin edges
    impulse = np.histogram(interactions,bins=time,density=False)
    
    # run decision model
    for t in xrange(len(time)-1):
        pred[t+1] = pred[t] - (pred[t] - c)/tau + sigma*noise[t] + dx*impulse[0][t]
        
    decision = time[np.argwhere(pred >= 1)] + sigma2*np.random.randn(1)
    if decision.shape[0] == 0:
        decision = max(time) + sigma2*np.random.randn(1)
    elif len(decision) > 1:
        decision = decision[0]
    
    return [decision, pred]
```

```python
def integrateAndForage2(interactions,
                       timeOfDeparture,
                       c = 0.01, # internal bias to forage
                       #dx = 0.2, # effect size of interaction
                       tau = 10, # time constant of ant's memory
                       sigma = 0.05, # input noise variance
                       sigma2 = 0,
                       threshold = 1): # output noise variance
    # interactions needs to be numpy array
    interactions = np.array(interactions)
    
    # initialize constants and variables
    penalty = 100   # penalty for not deciding to forage
    dt      = 1.0/30.0
    
    if interactions.shape[0] > 0:
        time    = np.linspace(min(interactions),timeOfDeparture,(timeOfDeparture - min(interactions))/dt)
    else:
        time    = np.linspace(0,timeOfDeparture,timeOfDeparture/dt)      
    pred    = np.zeros(len(time))
    noise   = np.random.randn(len(time))
    pred[0] = sigma*noise[0]
    
    # discretize interactions into our sample rate
    # impulse[0] are number of samples, impulse[1] are bin edges
    impulse = np.histogram(interactions,bins=time,density=False)
    
    # run decision model
    for t in xrange(len(time)-1):
        pred[t+1] = pred[t] - (pred[t] - c)/tau + sigma*noise[t] + impulse[0][t]
        
    decision = time[np.argwhere(pred >= 1)] + sigma2*np.random.randn(1)
    if decision.shape[0] == 0:
        decision = max(time) + sigma2*np.random.randn(1) + penalty
    elif len(decision) > 1:
        decision = decision[0]
    
    return [decision, pred]
```

#### Error models for full and reduced models

```python
def modelError(params, interactionTimes, timeOfDeparture):
    '''returns mean square error of integrate-and-forage model'''
    
    
    # error for each ant
    errors = np.zeros(len(interactionTimes))
    
    for i in xrange(len(interactionTimes)):
        pred = integrateAndForage(interactionTimes[i],
                                  timeOfDeparture[i],
                                  c      = params[0],
                                  dx     = params[1],
                                  tau    = params[2],
                                  sigma  = params[3],
                                  sigma2 = params[4])
        
        # use last interaction as time decision was made
        if interactionTimes[i].shape[0] > 0:
            errors[i] = (pred[0] - interactionTimes[i][-1])**2
        else:
            errors[i] = (pred[0] - timeOfDeparture[i])**2
    
    return sum(errors)/np.float(len(errors))
```

```python
def modelError2(params, interactionTimes, timeOfDeparture):
    '''returns mean square error of integrate-and-forage model'''
    
    
    # error for each ant
    errors = np.zeros(len(interactionTimes))
    
    for i in xrange(len(interactionTimes)):
        pred = integrateAndForage2(interactionTimes[i],
                                  timeOfDeparture[i],
                                  c      = params[0],
                                  #dx     = params[1],
                                  tau    = params[1],
                                  sigma  = params[2],
                                  sigma2 = params[3])
        
        # use last interaction as time decision was made
        if interactionTimes[i].shape[0] > 0:
            errors[i] = (pred[0] - interactionTimes[i][-1])**2
        else:
            errors[i] = (pred[0] - timeOfDeparture[i])**2
    
    return sum(errors)/np.float(len(errors))
```
