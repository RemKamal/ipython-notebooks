```python
def center_surround(numPoints):
    x = randn(10e4)
    y = randn(10e4)
    H,xedges,yedges = np.histogram2d(x,y,int(numPoints/2.0))
    H2,xedges2,yedges2 = np.histogram2d(x,y,2*numPoints)
    H2 = H2[int(numPoints/2.0):int(3.0*numPoints/2.0),int(numPoints/2.0):int(3.0*numPoints/2.0)]
    
    H_small = np.zeros((numPoints,numPoints))
    H_small[int(numPoints/4.0):int(3.0*numPoints/4.0),int(numPoints/4.0):int(3.0*numPoints/4.0)] = H
    
    spatial_filter2d = H_small - (5*H2-2*H_small)
    
    exp_filter = np.zeros((20,20))
    for i in xrange(exp_filter.shape[0]):
        exp_filter[i,:] = linspace(0,1,20)
    for j in xrange(exp_filter.shape[1]):
        exp_filter[:,j] = linspace(exp_filter[0,j],1,20)

    exp_filter = exp(-exp_filter)
    spatial_filter2d = ndimage.convolve(spatial_filter2d, exp_filter)
    spatial_filter2d = spatial_filter2d/np.max(spatial_filter2d)
    
    return spatial_filter2d - np.mean(spatial_filter2d)
```

```python
def monophasic(numPoints):
    x = randn(10e4)
    y = randn(10e4)
    H,xedges2,yedges2 = np.histogram2d(x,y,2*numPoints)
    H = H[int(numPoints/2.0):int(3.0*numPoints/2.0),int(numPoints/2.0):int(3.0*numPoints/2.0)]
    
    exp_filter = np.zeros((20,20))
    for i in xrange(exp_filter.shape[0]):
        exp_filter[i,:] = linspace(0,1,20)
    for j in xrange(exp_filter.shape[1]):
        exp_filter[:,j] = linspace(exp_filter[0,j],1,20)

    exp_filter = exp(-exp_filter)
    spatial_filter2d = ndimage.convolve(H, exp_filter)
    spatial_filter2d = spatial_filter2d/np.max(spatial_filter2d)
    
    return spatial_filter2d #- np.mean(spatial_filter2d)
```

## Construct stimuli

#### Sunrise/sunset experiment

```python
def sunrise(cellsPerSide=4, slope=1.0):
    frames  = int(255.0/slope)
    randn1d = (255 - 127)*0.15*randn(1,1,frames)
    randn41 = np.row_stack(2*(randn1d,randn1d))
    stim    = np.hstack(int(cellsPerSide/2.0)*(randn41,randn41)) # Uniform flicker
    
    for t in xrange(frames):
        mean = t*1.0
        stim[:,:,t] = stim[:,:,t] + mean
        
    stim = stim.astype(int)
    stim[stim > 255] = 255
    stim[stim < 0]   = 0
    
    return stim
```

```python
stim = sunrise()
plot(stim[0,0,:])
xlim((0,stim.shape[2]))
ylim((0,255))
```
