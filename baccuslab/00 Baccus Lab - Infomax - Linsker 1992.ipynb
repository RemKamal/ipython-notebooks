```python
def varianceConstraint(C,Q,eta):
    allNodes = diag((C.dot(Q)).dot(C.transpose()) + eta*C.dot(C.transpose()))
    return sum([abs(1 - v) for v in allNodes])
```

```python
def varianceFreqConstraint(q,c):
    if len(q)!=len(c):
        print 'q and c are not the same length!'
    else:
        K = len(q)
        return (1./K)*sum([(q[k] + eta)*(c[k]**2) for k in range(K)])
```

```python
def callbackFn(mu):
    #print varianceFnMu(mu)
    print mu
```

```python
def uniqueSolution(eta,beta,q,mode='manual'):
    K = len(q)
    
    def getFilter(mu):
        return np.array([np.sqrt((beta/eta)*max([0,(-1 + (qk/(2*(qk+eta)))*(1 + np.sqrt(1-(2*eta*K)/(mu*beta*qk))))])) for qk in q])
    
    if mode=='auto':
        def varianceFnMu(mu):
            c = getFilter(mu)
            return 10*abs(1. - varianceFreqConstraint(q,c))

        mu  = -abs(randn(1))
        res = minimize(varianceFnMu, x0=mu, method='Nelder-Mead', callback=callbackFn)

        muOpt = res.x
        return getFilter(muOpt)
    else:
        return getFilter
```

```python
def manualGradientDescent(x, original, iterations=500):
    if len(x) == 6:
        #sigma1,sigma2,sigma3,strength1,strength2,strength3 = x
        errorFn = twoGaussiansError
    elif len(x) == 4:
        #sigma1,sigma2,strength1,strength2 = x
        errorFn = oneGaussiansError
    else:
        print 'Wrong number of initial parameters. Should be 4 or 6.'
        
    i      = 0
    step   = 0.5
    errors = []
    perturb  = randn(len(x))
    dfdtheta = np.zeros((len(x),))
    for g in range(len(x)):
        newX    = x
        newX[g] += perturb[g]
        dfdtheta[g] = (errorFn(newX, original) - errorFn(x, original))/perturb[g]
        
    while i < iterations:
        x -= step*dfdtheta   
        i += 1
        step = step/(5 + i)
        errors.append(errorFn(x, original))
        
        perturb = randn(len(x))
        for g in range(len(x)):
            newX    = x
            newX[g] += perturb[g]
            dfdtheta[g] = (errorFn(newX, original) - errorFn(x, original))/perturb[g]
        
    return x, errors   
```

```python
def fourGaussians(x,original):
    sigma1,sigma2,sigma3,sigma4,strength1,strength2,strength3,strength4 = x
    xs = linspace(-5,5,len(original))
    ourGaussian = abs(strength1)*gaussian(x=xs, sigma=abs(sigma1)) - abs(strength2)*gaussian(x=xs, sigma=abs(sigma2)) + abs(strength3)*gaussian(x=xs, sigma=abs(sigma3)) - abs(strength4)*gaussian(x=xs, sigma=abs(sigma4))
    return ourGaussian
```

```python
def fourGaussiansError(x,original):
    ourGaussian = fourGaussians(x,original)
    return sum(abs(ourGaussian-original))
```

```python
def fitGaussians(sigma1,sigma2,sigma3,sigma4,strength1,strength2,strength3,strength4):
    original=cOpt
    ourGaussian = fourGaussians([sigma1,sigma2,sigma3,sigma4,strength1,strength2,strength3,strength4],original)
    
    print sum(abs(ourGaussian-original))

    plot(original,'k',linewidth=3)
    plot(ourGaussian,'r',linewidth=2,alpha=0.8)
    plot(oneGaussian([sigma3,sigma4,strength3,strength4],original),'b',linewidth=1.5,alpha=0.8)
    plot(oneGaussian([sigma1,sigma2,strength1,strength2],original),'g',linewidth=1.5,alpha=0.8)
    
    return ourGaussian
    
```
