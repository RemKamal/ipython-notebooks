# generalized k-Nearest Neighbors entropy

```python
import numpy as np
from math import gamma
from sklearn.neighbors import DistanceMetric
```

```python
def entropy(data, ball='euclidean', k=1, units='nats'):
    """
    Estimates the entropy of the given data using the k-nearest neighbors method

    input
    -----
    data (nd-array):
        An (n by p) matrix containing n samples of p-dimensional data

    ball (string):
        Which ball (e.g. l1, euclidean, etc.) to use when computing the volume.
        Acceptable strings include:
            'l1'   : l1 or Manhattan distance
            'l2'   : l2 or Euclidean distance; default
            'linf' : l-infinity or Chebyshev distance

    k (integer):
        How many nearest-neighbors to use when computing radii. Must be at least 1.

    """
    
    # Get number of samples and dimensionality
    (n,p)  = data.shape
    
    # Determine radii and volumes for a given metric space
    metric = getball(ball)
    if metric == 1:
        m = 'manhattan'
    elif metric == 2:
        m = 'euclidean'
    elif metric == inf:
        m = 'chebyshev'
        
    dist  = DistanceMetric.get_metric(m)
    D_mat = dist.pairwise(data)
    D_mat.sort(axis=1)
    radii = D_mat[:,k]
    Vs    = volume(radii, ball=str(metric), dimension=p)
    
    if units.lower() == 'nats':
        return sum([np.log(vol) for vol in Vs])/float(n) + np.log(n) - L(k - 1) + 0.577215665
    if units.lower() == 'bits':
        return sum([np.log2(vol) for vol in Vs])/float(n) + np.log2(n) - L(k - 1) + 0.577215665
```

```python
def L(k):
    if k == 0:
        return 0
    elif k > 0:
        return sum([1/float(i) for i in xrange(1,k+1)])
```

```python
def volume(radii, ball, dimension):
    """
    Computes the volume of the given ball with given radius

    input
    -----
    radii (nd-array):
        A set of radius values to use when computing volume

    ball (string):
        Which ball (e.g. l1, euclidean, etc.) to use when computing the volume

    dimension (integer):
        Dimensionality of the space in which to compute the volume

    output
    ------
    volume (nd-array):
        A set of computed volumes, one for each given radius

    """
    p = float(dimension)
    b = getball(ball)
    return (2*gamma(1/b + 1)*radii)**p / gamma(p/b + 1)
```

```python
def getball(string):
    """
    Get index associated with a corresponding 'ball'

    input
    -----
    string:
        A given name of a 'ball' (e.g. euclidean, l1, l2, etc.)

    output
    ------
    index:
        An index (integer) associated with that ball:
        1: Manhattan (l1) ball
        2: Euclidean (l2) ball
        inf: Chebyshev (l_inf) ball

    """

    # Euclidean ball
    if string.upper() == 'L2' or string.upper() == 'EUCLIDEAN' or string == '2':
        return 2

    # l1 ball
    elif string.upper() == 'L1' or string.upper() == 'MANHATTAN' or string.upper() == 'TAXICAB' or string == '1':
        return 1

    elif string.upper() == 'LINF' or string.upper() == 'CHEBYSHEV' or string == 'inf':
        return float('inf')

    # ball not found
    else:
        raise ValueError('Could not recognize the ' + string + ' ball.')
```

```python
def meanEntropy(data, ball='l2', units='nats'):
    """
    Estimates the entropy of the given data using the mean k-nearest neighbors method
    from Faivshevsky and Goldberger that appeared at NIPS.

    input
    -----
    data (nd-array):
        An (n by p) matrix containing n samples of p-dimensional data

    ball (string):
        Which ball (e.g. l1, euclidean, etc.) to use when computing the volume.
        Acceptable strings include:
            'l1'   : l1 or Manhattan distance
            'l2'   : l2 or Euclidean distance; default
            'linf' : l-infinity or Chebyshev distance

    k (integer):
        How many nearest-neighbors to use when computing radii. Must be at least 1.

    """
    
    # Get number of samples and dimensionality
    (n,p)  = data.shape
    
    # Determine radii and volumes for a given metric space
    metric = getball(ball)
    if metric == 1:
        m = 'manhattan'
    elif metric == 2:
        m = 'euclidean'
    elif metric == inf:
        m = 'chebyshev'
        
    dist  = DistanceMetric.get_metric(m)
    D_mat = dist.pairwise(data)
    D_mat.sort(axis=1)
    H = []
    for k in xrange(1,n-1):  
        radii = D_mat[:,k]
        Vs    = volume(radii, ball=str(metric), dimension=p)
        if units.lower() == 'nats':
            H.append(sum([np.log(vol) for vol in Vs])/float(n) + np.log(n) - L(k - 1) + 0.577215665)
        elif units.lower() == 'bits':
            H.append(sum([np.log2(vol) for vol in Vs])/float(n) + np.log2(n) - L(k - 1) + 0.577215665)
        
    return np.mean(H)
```

## Simulations

#### Gaussian white noise

```python
from numpy.random import randn
```

```python
n = 10e3
k = range(1,int(n-1))
b = ['l1', 'l2', 'linf']
p = 3

data  = randn(n,p)
Sigma = cov(data.transpose())

true_entropy = (p/2.0)*(1 + np.log(2*pi)) + (1/2.0)*np.log(det(Sigma))
est_entropy  = np.zeros((len(k),len(b)))
for idx, kth in enumerate(k):
    print idx
    for jdx, ball in enumerate(b):
        est_entropy[idx,jdx] = entropy(data,ball,kth,units='nats')
```

```python
true_entropy
```

```python
l1_entropies = [est_entropy[i,0] for i in xrange(0,est_entropy.shape[0]) if est_entropy[i,0] > 0]
l2_entropies = [est_entropy[i,1] for i in xrange(0,est_entropy.shape[0]) if est_entropy[i,1] > 0]
linf_entropies = [est_entropy[i,2] for i in xrange(0,est_entropy.shape[0]) if est_entropy[i,2] > 0]
```

```python
mean(l1_entropies)
```

```python
std(l1_entropies)
```

```python
fig = gcf()
fig.set_size_inches((12,8))

ks = range(1,len(l1_entropies)+1)
plot(ks, l1_entropies, 'b')
plot(ks, l2_entropies, 'k')
plot(ks, linf_entropies, 'r')
plot(ks, [true_entropy for k in ks], 'k--')
xlim((1,max(ks)))
ylim((4.1,4.8))
xlabel('k nearest neighbor (10e3 total samples)', fontsize=16)
ylabel('Entropy estimate of a 3d Gaussian (Nats)', fontsize=16)
title('Comparison of gkNN entropy estimators for 3d Gaussian', fontsize=18)
annotate('True entropy of 3d Gaussian', (500, 4.3), fontsize=14)
annotate('Estimated entropy using L2 distance', (100, 4.62), fontsize=14)
annotate('Estimated entropy using L1 (blue) and Chebyshev (red) distance', (200, 4.15), fontsize=14)

savefig('Comparison of gkNN entropy estimators for 3d Gaussian.png')
```
