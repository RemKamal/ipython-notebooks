#### Functions to get predictive filter and infomax filter

```python
def getW(x,y,mode='auto'):
    ''' Solve XW = Y where X is toeplitz, y is a vector.
    
    1) 'auto' mode uses numpy's builtin lstsq solver to avoid
    taking matrix inversions or multiplying by illconditioned matrices.
    2) 'direct' mode uses direct matrix inversion by xtx and returns a 
    warning when xtx is ill-conditioned.
    '''
    
    if mode=='direct':
        xtx = np.dot(np.transpose(x),x)
        con = np.linalg.cond(xtx)
        if con > 10e3:
            print('Condition number of XTX is very high, %f, and inverse wont be accurate.') % (con)
        W = np.dot(np.linalg.inv(xtx), np.dot(np.transpose(x),y))
    elif mode=='auto':
        W, res, rank, s = np.linalg.lstsq(x, y, rcond=-1)
    else:
        print('mode can be either auto or direct')
    
    return W
```

```python
def toe(x,height,width):
    '''x is a vector, height is dimension m = # of predictions,
    and width is dimension n = length of filter.
    '''
    m = height
    n = width
    first_row = x[:n]
    first_row = first_row[::-1] #reverse the first row, since it's x_t, x_t-1, ..., x_t-n+1
    first_col = x[n-1:]  #this should go from n to m but it's weird since python doesn't count the first element

    return toeplitz(first_col,first_row)
```

```python
def gaussianInfoSNR(w,x,x_in,mode='variance'):
    if mode=='variance':    
        S = np.var(np.convolve(x,w,'valid'))
        N = np.var(np.convolve(x_in-x,w,'valid'))
        return 0.5*np.log2(1+S/N)
    
    elif mode=='power-spectrum':
        signal = np.convolve(x,w,'valid')
        noise  = np.convolve(x_in-x,w,'valid')
        n = len(signal) # length of the signal
        S = fft(signal)/n # fft computing and normalization #-np.mean(signal)
        S = abs(S[range(n/2)])
        S = np.convolve(S,(1.0/300.0)*np.ones(500),'same')
        N = fft(noise)/n
        N = abs(N[range(n/2)])
        N = np.convolve(N,(1.0/300.0)*np.ones(500),'same')
    
        # SNR(f) is P_s/P_n or (A_s/A_n)**2
        #SNR = [(S[f]/N[f])**2 for f in xrange(n/2)] => this is what van Hateren does and it's wrong.
    
        return 0.5*np.log2(1+(np.mean(S)/np.mean(N))**2)
```

```python
def negInfo_SNR(w,x,x_in):
    return -gaussianInfoSNR(w,x,x_in)
```

```python
def negInfo_SNR_smoothed(w,xs,x_ins):
    mi = []
    for t in xrange(len(xs)):
        mi.append(-gaussianInfoSNR(w,xs[t],x_ins[t]))
        
    return np.mean(mi)
```

```python
def gaussianInfoSNR_bothNoise(w,x,inoise,onoise,mode='variance'):
    if mode=='variance':    
        S = np.var(np.convolve(x,w,'valid'))
        N = np.var(np.convolve(inoise,w,'valid') + onoise)
        return 0.5*np.log2(1+S/N)
    
    elif mode=='power-spectrum':
        signal = np.convolve(x,w,'valid')
        noise  = np.convolve(x_in-x,w,'valid')
        n = len(signal) # length of the signal
        S = fft(signal)/n # fft computing and normalization #-np.mean(signal)
        S = abs(S[range(n/2)])
        S = np.convolve(S,(1.0/300.0)*np.ones(500),'same')
        N = fft(noise)/n
        N = abs(N[range(n/2)])
        N = np.convolve(N,(1.0/300.0)*np.ones(500),'same')
    
        # SNR(f) is P_s/P_n or (A_s/A_n)**2
        #SNR = [(S[f]/N[f])**2 for f in xrange(n/2)] => this is what van Hateren does and it's wrong.
    
        return 0.5*np.log2(1+(np.mean(S)/np.mean(N))**2)
```

```python
def negInfo_SNR_bothNoise(w,x,inoise,onoise):
    return -gaussianInfoSNR_bothNoise(w,x,inoise,onoise)
```

```python
def powerConstraint(w,x_in,onoise):
    return 1.0 - (np.var(np.convolve(x_in,w)) + np.var(onoise)) # since variance of independent variables sum
```

```python
def plotSpectrum(y,Fs=1000.0,colorstr='r',fillin=True,transparency=0.7,plt=True):
 '''
 Plots a Single-Sided Amplitude Spectrum of y(t).
 
 Example:
 Fs = 1000.0;  # sampling rate
 Ts = 1.0/Fs; # sampling interval
 t = arange(0,len(W)*Ts,len(W)) # time vector

 subplot(2,1,1)
 plot(W, 'k')
 xlim(-5,n+5)
 title('W', fontsize=14)
 ax = gca()
 ax.set_xlim(ax.get_xlim()[::-1])
 xlabel('Time before present', fontsize=14)

 subplot(2,1,2)
 plotSpectrum(W,Fs)
 '''
 n = len(y) # length of the signal
 k = arange(n)
 T = n/Fs
 frq = k/T # two sides frequency range
 frq = frq[range(n/2)] # one side frequency range

 Y = fft(y)/n # fft computing and normalization
 Y = Y[range(n/2)]
 
 if plt:
     plot(frq,abs(Y),color=colorstr,alpha=transparency) # plotting the spectrum
     if fillin:
            fill_between(frq, abs(Y), 1e-6, color=colorstr, alpha=0.4)
     xlabel('Freq (Hz)', fontsize=14)
     ylabel('|Y(freq)|', fontsize=14)
     xlim(-10,frq[-1])
 else:
     return frq, abs(Y)
```

```python
def smoothData(x,y,numPoints=1000.,variance=1000.):
    window = scipy.signal.gaussian(numPoints, variance)
    smooth_data = np.convolve(y,window/window.sum(),mode='valid')
    smooth_x  = np.linspace(x[0],x[-1],len(smooth_data))
    
    interpolation = interp1d(smooth_x,smooth_data)
    y_HD = interpolation(x)
    
    return y_HD
```

```python
def smoothEnvelopePlot(x,y,colorstr='k',transparency=0.2,plt=False,linesty='-',nPoints=1000.,var=1000.):

    y_HD = smoothData(x,y,numPoints=nPoints,variance=var)

    error = [np.sqrt((y_HD[t]-y[t])**2) for t in xrange(len(x))]
    err   = smoothData(x,error,numPoints=nPoints,variance=var)
    
    if plt:
        fill_between(x,y_HD-err,y_HD+err,color=colorstr,alpha=transparency)
        plot(x,y_HD,color=colorstr,linestyle=linesty,linewidth=2)
    else:
        return y_HD, err
```

```python
def autocorr(x,y):
    result = np.correlate(x, y, mode='full')
    return result[result.size/2:]
```

```python
def polyfit_line(x,y,degree=1):
    if type(x) is list:
        x = array(x) # ** doesn't support lists
    ps = np.polyfit(x,y,deg=degree)
    for idx,p in enumerate(ps):
        if idx == 0:
            y = p*x**(degree-idx)
        elif idx == degree:
            y = y + p
        else:
            y = y + p*x**(degree-idx)
    return y
```

```python
def pinkNoise(duration):
    x_t    = np.random.randn(2*duration)
    x_f    = np.fft.rfft(x_t)[:len(x_t)/2]
    f      = linspace(1,duration,len(x_f))
    fi     = 1./f
    pink_f = x_f*fi

    return real(np.fft.ifft(pink_f))
```

```python
from scipy.optimize import minimize
```

```python
def error(coeff, X_past_nonGauss, w_nonGauss, y_nonGauss):
    cutoff = coeff[0]
    scale  = coeff[1]
    
    predicted_nonGauss = (np.dot(X_past_nonGauss,w_nonGauss))
    for idx,p in enumerate(predicted_nonGauss):
        if p > cutoff:
            predicted_nonGauss[idx] = p*scale
            
    return sum(np.abs(y_nonGauss - predicted_nonGauss)**2)
```

```python
def numPts(x,start,stop):
    selected = [p for p in x if p > start and p <= stop]
    return len(selected)
```

```python
def myrange(x):
    return np.max(x)-np.min(x)
```

Use polyfit to extrapolate out to maximum predicted value

```python
def polyfit_line(x,y,newx=x,degree=1):
    if type(x) is list:
        x = array(x) # ** doesn't support lists
    ps = np.polyfit(x,y,deg=degree)
    x = newx
    for idx,p in enumerate(ps):
        if idx == 0:
            y = p*x**(degree-idx)
        elif idx == degree:
            y = y + p
        else:
            y = y + p*x**(degree-idx)
    return y
```

Polynomial fit doesn't work well for extrapolating.  Try fitting a sigmoid instead

```python
def sigmoid(params, x):
    gain = params[0]
    threshold = params[1]
    k = params[2]
    #lowerbound = params[3]
    return k/(1+np.exp(-gain*(x - threshold))) # - lowerbound, this parameter destroys generalization
```

```python
def meanSquareError(y1,y2):
    if type(y1) is list:
        y1 = array(y1)
    if type(y2) is list:
        y2 = array(y2)
    return np.mean((y1-y2)**2)
```

```python
def errorSigmoid(params, x, y):
    y_fit = sigmoid(params, x)
    return meanSquareError(y_fit, y)
```

```python
def fitSigmoid(x,y,xnew=x):
    coeff = randn(3)
    res   = minimize(errorSigmoid, x0=coeff, args=(x,y))
    return sigmoid(res.x,xnew)
```
